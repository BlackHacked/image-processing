# 图像二值化函数

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2GRAY)
```

```python
def tobinary(gray,thresh):
    dst = np.zeros_like(gray)
    idx = (gray >= thresh)
    dst[idx] = 255
    return dst

def tobinary_min_max(gray,minval,maxval):
    dst = np.zeros_like(gray)
    df_gray = DataFrame(gray)
    minval = min(np.abs(minval),np.abs(maxval))
    maxval = max(np.abs(minval),np.abs(maxval))
    idx = (df_gray >= minval) & (df_gray <= maxval)
    dst[idx] = 255
    return dst
```

```python
dst = tobinary(src,150)
```

```python
plt.subplot(121)
plt.imshow(src,'gray')
plt.title('Gray')
plt.subplot(122)
plt.imshow(dst,'gray')
plt.title('Binary')
```

```python
dst = tobinary_min_max(src,150,255)
```

```python
plt.subplot(121)
plt.imshow(src,'gray')
plt.title('Gray')
plt.subplot(122)
plt.imshow(dst,'gray')
plt.title('Binary')
```
