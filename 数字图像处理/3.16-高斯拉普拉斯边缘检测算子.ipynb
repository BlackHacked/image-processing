# 高斯拉普拉斯边缘检测算子

图像边缘可表示为一阶过极点或二阶过零点. 
拉普拉斯算子表示为$d2f = a2f / a2x + a2f / a2y $
由于噪声点对边缘检测影响较大，所以由高斯滤波和拉普拉斯算子结合，形成高斯拉普拉斯算子。 
由高斯函数根据拉普拉斯算子的公式求导可得 
![](doc_imgs/01.jpg)

离散化，取sigma为1.0，则5x5模块为 

{0.0175 0.0392 0.0431 0.0392 0.0175 

0.0392 0 -0.0965 0 0.0392 

0.0431 -0.0965 -0.3183 -0.0965 0.0431 

0.0392 0 -0.0965 0 0.0392 

0.0175 0.0392 0.0431 0.0392 0.0175} 

对其取整，并使总和为0，则 

{ -2, -4, -4, -4, -2, 
-4, 0, 8, 0, -4, 
-4, 8, 24, 8, -4, 
-4, 0, 8, 0, -4, 
-2, -4, -4, -4, -2 };

```python
#coding:utf-8
%matplotlib inline
import numpy as np
import scipy
from scipy.special import comb
import matplotlib.pyplot as plt
import cv2
from scipy import stats
```

```python
src = cv2.imread('../datas/f4.jpg')
```

```python
kernel = np.array([-2, -4, -4, -4, -2, 
-4, 0, 8, 0, -4, 
-4, 8, 24, 8, -4, 
-4, 0, 8, 0, -4, 
-2, -4, -4, -4, -2]).reshape((5,5))
```

```python
dst = cv2.filter2D(src,-1,kernel)
```

```python
plt.imshow(dst)
```
