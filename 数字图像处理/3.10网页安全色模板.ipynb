```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
from scipy.misc import imread, imresize, imsave
```

```python
src = cv2.cvtColor(cv2.imread('../datas/rgb-cube.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_RGB2GRAY)
```

```python
dst = np.fix((src.astype(np.float32) / 51.0) * 51.0).astype(np.uint8)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('Original')
plt.subplot(122)
plt.imshow(dst)
plt.title('Web Colors')
```
