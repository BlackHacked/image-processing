# 图像形态-TopHat和BottomHat

```python
from skimage import data,color
%matplotlib inline
import matplotlib.pyplot as plt
from skimage.morphology import disk
import skimage.io
import skimage.filters.rank as sfr
import numpy as np
```

```python
src = skimage.io.imread('../datas/f2.jpg')
gray =color.rgb2gray(src)
```

```python
auto =sfr.bottomhat(gray, disk(5)) #半径为5的圆形滤波器
```

```python
plt.imshow(auto,'gray')
plt.title('BottomHat')
```

```python
auto =sfr.tophat(gray, disk(5)) #半径为5的圆形滤波器
```

```python
plt.imshow(auto,'gray')
plt.title('TopHat')
```
