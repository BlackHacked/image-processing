# 灰度拉伸算法

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2GRAY)
```

```python

def gethistogram(src):
    hist = np.zeros((256,),dtype=np.int64)
    temp = np.reshape(src,(src.shape[0] * src.shape[1])).tolist()
    for i in range(256):
        hist[i] = temp.count(i)
    return hist

def gray_stretch(src):
    min_val = np.min(src)
    max_val = np.max(src)
    if min_val == max_val:
        return src.oopy()
    dst = float(255.0 / (max_val - min_val)) * (np.float64(src) - min_val)
    return cv2.convertScaleAbs(dst)
```

```python
dst = gray_stretch(src)
hist = gethistogram(dst)
hist_stretch = gethistogram(dst)
```

```python
plt.subplot(121)
plt.imshow(src,'gray')
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst,'gray')
plt.title('Gray Stretch')
```

```python
plt.subplot(121)
plt.bar(np.arange(0,256),hist)
plt.title('Histogram:SRC')
plt.subplot(122)
plt.bar(np.arange(0,256),hist_stretch)
plt.title('Histogram:Stretch')
```
