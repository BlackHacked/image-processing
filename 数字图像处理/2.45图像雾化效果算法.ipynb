# 图像雾化效果算法

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def atomization_effect(src,havy):
    rows,cols = src.shape[0],src.shape[1]
    k = 0
    dx = 0
    dy = 0
    dst = src.copy()
    for i in range(rows):
        for j in range(cols):
            kx = np.random.randint(havy)
#             ky = np.random.randint(havy)
            dx = j + kx
            if dx >= cols:
                dx = cols - 1
            dy = i + kx
            if dy >= rows:
                dy = rows - 1
                
            dst[i,j] = src[dy,dx]
            
    return dst
    
```

```python
dst = atomization_effect(src,16)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Atomization Effect')
```
