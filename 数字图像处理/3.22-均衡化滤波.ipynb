# 均衡化滤波

利用局部直方图对图像进行均衡化滤波

```python
from skimage import data,color
%matplotlib inline
import matplotlib.pyplot as plt
from skimage.morphology import disk
import skimage.io
import skimage.filters.rank as sfr
import numpy as np
```

```python
src = skimage.io.imread('../datas/f2.jpg')
gray =color.rgb2gray(src)
```

```python
dst =sfr.equalize(gray, disk(5)) #半径为5的圆形滤波器
```

```python
plt.imshow(dst,'gray')
```
