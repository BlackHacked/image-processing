# 图像局部熵

求局部熵，熵是使用基为2的对数运算出来的。该函数将局部区域的灰度值分布进行二进制编码，返回编码的最小值。

```python
from skimage import data,color
%matplotlib inline
import matplotlib.pyplot as plt
from skimage.morphology import disk
import skimage.io
import skimage.filters.rank as sfr
import numpy as np
```

```python
src = skimage.io.imread('../datas/f2.jpg')
gray =color.rgb2gray(src)
```

```python
dst =sfr.entropy(gray, disk(5)) #半径为5的圆形滤波器
```

```python
plt.imshow(dst,'gray')
```
