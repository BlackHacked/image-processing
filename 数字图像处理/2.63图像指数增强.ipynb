# 图像指数增强

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def index_enhance(src):
    rgb = np.float64(src)
    dst = np.zeros_like(rgb)
    dst = 1.0 / 255 * rgb * rgb
    dst[dst < 0] = 0
    dst[dst > 255] = 255
    return cv2.convertScaleAbs(dst)
```

```python
dst = index_enhance(src)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('DST')
```
