# 图像马赛克效果算法

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def mosaic(src,v):
    rows,cols = src.shape[0],src.shape[1]
#     dst = np.zeros_like(src)
    dst = src.copy()
    
    dR = src[0,0,0]
    dG = src[0,0,1]
    dB = src[0,0,2]
    dstX = 0
    dstY = 0
    
    for i in range(1,rows):
        for j in range(1,cols):
            dstX = j
            dstY = i
            
            if i % v == 0:
                if j % v == 0:
                    dR = src[dstY,dstX,0]
                    dG = src[dstY,dstX,1]
                    dB = src[dstY,dstX,2]
                else:
                    dst[dstY,dstX,0] = dR
                    dst[dstY,dstX,1] = dG
                    dst[dstY,dstX,2] = dB
            else:
                dst[dstY,dstX,0] = dst[dstY - 1,dstX,0] 
                dst[dstY,dstX,1] = dst[dstY - 1,dstX,1] 
                dst[dstY,dstX,2] = dst[dstY - 1,dstX,2]
                
    return dst
    
```

```python
dst = mosaic(src,16)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Mosaic')
```
