# 图像亮度调整

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def brightness_adjust(src,brightness):
    dst = np.float64(src) + brightness
    dst[dst > 255] = 255
    dst[dst < 0] = 0
    return cv2.convertScaleAbs(dst)
```

```python
dst = brightness_adjust(src,-64) 
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('RGB')
plt.subplot(122)
plt.imshow(dst)
plt.title('Brightness')
```
