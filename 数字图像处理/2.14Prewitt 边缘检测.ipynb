# Prewitt 边缘检测

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def prewitt_edge(src):
    rgb = np.float64(src)
    dst = np.zeros_like(rgb)
    rows,cols = rgb.shape[0],rgb.shape[1]
    
    for x in range(1,rows - 1):
        for y in range(1,cols - 1):
            pixel = np.abs(rgb[x + 1,y - 1] + rgb[x + 1,y] + rgb[x + 1,y + 1] - \
                           rgb[x - 1,y - 1] - rgb[x - 1,y] - rgb[x - 1,y + 1]) + \
                    np.abs(rgb[x - 1,y - 1] + rgb[x,y - 1] + rgb[x + 1,y - 1] - \
                           rgb[x - 1,y + 1] - rgb[x,y + 1] - rgb[x + 1,y + 1])
            dst[x,y] = pixel
    
    dst[dst < 0] = 0
    dst[dst > 255] = 255
    
    return cv2.convertScaleAbs(dst)
```

```python
dst = prewitt_edge(src)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('RGB')
plt.subplot(122)
plt.imshow(dst)
plt.title('Prewitt Edge')
```

```python
gray = cv2.cvtColor(src,cv2.COLOR_RGB2GRAY)
dst = prewitt_edge(gray)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('RGB')
plt.subplot(122)
plt.imshow(dst,'gray')
plt.title('Prewitt Edge')
```
