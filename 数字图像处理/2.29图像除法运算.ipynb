# 图像除法运算

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
src2 = cv2.cvtColor(cv2.imread('../datas/f1.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def division_operate(src1,src2):
    s1 = np.array(src1.shape)
    s2 = np.array(src2.shape)
    dst = np.float64(src1.shape,dtype=np.float64)
    res = s1 == s2
    if not res.all():
        raise ValueError('src1.shape != src2.shape')
    eps = 1.0e-5
    
    dst = src1 * np.abs(np.float64(src1) - np.float64(src2)) / (src1 + src2 + eps)
    
    return cv2.convertScaleAbs(dst)
```

```python
dst = division_operate(src,src2)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('RGB')
plt.subplot(122)
plt.imshow(dst)
plt.title('Division')
```
