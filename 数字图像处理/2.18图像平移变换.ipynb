# 图像平移变换

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def translation(src,x,y):
    rows,cols = src.shape[0],src.shape[1]
    dst = np.zeros_like(src)
    if x < 0:
        x = 0
    if x > cols:
        x = cols
    if y < 0:
        y = 0
    if y > rows:
        y = rows
    
    dst[y:rows,x:cols] = src[0:rows - y,0:cols - x]
    
    return dst
```

```python
dst = translation(src,50,50)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('RGB')
plt.subplot(122)
plt.imshow(dst)
plt.title('Translation')
```
