# 图像对比度调整

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def contrast_adjust(src,contrast):
    # 整个图片的平均亮度
    mean_val = np.mean(src)
    dst = np.zeros(src.shape,dtype=np.float64)
    
    dst = (np.float64(src) - mean_val) * contrast + np.float64(src)
    dst[dst > 255] = 255
    dst[dst < 0] = 0
    
    return cv2.convertScaleAbs(dst)
```

```python
dst = contrast_adjust(src,1.2) 
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('RGB')
plt.subplot(122)
plt.imshow(dst)
plt.title('Contrast')
```
