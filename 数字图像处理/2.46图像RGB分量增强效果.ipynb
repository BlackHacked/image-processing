# 图像RGB分量增强效果 

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def rgb_adjust(src,t):
    dst = np.float64(src) + t
    dst[dst < 0] = 0
    dst[dst > 255] = 255
    return cv2.convertScaleAbs(dst)
```

```python
dst = rgb_adjust(src,32)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('RGB Adjust')
```
