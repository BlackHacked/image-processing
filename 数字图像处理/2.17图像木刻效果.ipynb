# 图像木刻效果

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def woodcut_effect(src):
    rgb = np.float64(src)
    dst = np.zeros_like(rgb)
    
    rows,cols = rgb.shape[0],rgb.shape[1]
    
    for i in range(rows):
        for j in range(cols):
            avg = np.mean(rgb[i,j])
            if avg > 122.5:
                dst[i,j] = 0
            else:
                dst[i,j] = 255
                
    return cv2.convertScaleAbs(dst)
```

```python
dst = woodcut_effect(src)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('RGB')
plt.subplot(122)
plt.imshow(dst)
plt.title('WoodCut Effect')
```

```python
gray = cv2.cvtColor(src,cv2.COLOR_RGB2GRAY)
dst = woodcut_effect(gray)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('RGB')
plt.subplot(122)
plt.imshow(dst,'gray')
plt.title('WoodCut Effect')
```
