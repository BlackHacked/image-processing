# 图像颜色聚类

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def color_cluster(src,value):
    dst = np.bitwise_and(src,value)
    return dst
```

```python
dst = color_cluster(src,192) # 移6位
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('RGB')
plt.subplot(122)
plt.imshow(dst)
plt.title('Color Cluster')
```
