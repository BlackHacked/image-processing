# 图像非线性变换

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def nonlinear_transform(src,k):
    assert k >= 0 and k <= 128
    dst = np.zeros(src.shape,dtype=np.float64)
    rgb = np.float64(src)
    
    dst = k * np.log10(1 + rgb)
    dst[dst < 0] = 0
    dst[dst > 255] = 255
    
    return cv2.convertScaleAbs(dst)
```

```python
dst = nonlinear_transform(src,110)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Nonelinear Transform')
```
