# 自动色阶

```python
from skimage import data,color
%matplotlib inline
import matplotlib.pyplot as plt
from skimage.morphology import disk
import skimage.io
import skimage.filters.rank as sfr
import numpy as np
```

```python
src = skimage.io.imread('../datas/f2.jpg')
```

```python
img =color.rgb2gray(src).astype(np.float64)
auto =sfr.autolevel(img, disk(5)) #半径为5的圆形滤波器
```

```python
plt.imshow(auto,'gray')
```

```python
dst = np.zeros_like(src).astype(np.float64)
dst[:,:,0] = sfr.autolevel(src[:,:,0], disk(5))
dst[:,:,1] = sfr.autolevel(src[:,:,1], disk(5))
dst[:,:,2] = sfr.autolevel(src[:,:,2], disk(5))

dst = np.clip(dst,0,255)
dst = np.uint8(dst)
```

```python
plt.imshow(dst)
```
