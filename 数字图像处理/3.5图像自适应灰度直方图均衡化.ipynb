# 图像自适应灰度直方图均衡化

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
from scipy.misc import imread, imresize, imsave
```

```python
src = cv2.cvtColor(cv2.imread('../datas/fog2.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_RGB2GRAY)
```

```python
# 归一化
mri_max = np.amax(src)  
mri_min = np.amin(src)  
mri_img = ((src-mri_min)/(mri_max-mri_min))*255  
mri_img = mri_img.astype('uint8')  
```

```python
rows,cols,ch = src.shape
dst = np.zeros_like(src)
for k in range(ch):
    temp = mri_img[:,:,k]
    clahe = cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
    img = clahe.apply(temp)
    dst[:,:,k] = img
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Adative Histogram Equation')
```
