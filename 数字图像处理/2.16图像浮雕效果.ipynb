图像浮雕效果

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def relief_effect(src,relief):
    rgb = np.float64(src)
    dst = np.zeros_like(rgb)
    
    rows,cols = rgb.shape[0],rgb.shape[1]
    
    if relief < 0:
        relief = 0
    if relief > 255:
        relief = 255
        
    for x in range(0,rows - 1):
        for y in range(0,cols):
            pixel = np.abs(rgb[x,y] - rgb[x + 1,y] + relief)
            dst[x,y] = pixel
            
    dst[dst > 255] = 255
    
    return cv2.convertScaleAbs(dst)
```

```python
dst = relief_effect(src,64)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('RGB')
plt.subplot(122)
plt.imshow(dst)
plt.title('Relief Effect')
```

```python
gray = cv2.cvtColor(src,cv2.COLOR_RGB2GRAY)
dst = relief_effect(gray,64)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('RGB')
plt.subplot(122)
plt.imshow(dst,'gray')
plt.title('Relief Effect')
```
