# 图像垂直镜像

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def mirror_y(src):
    dst = np.zeros_like(src)
    rows,cols = src.shape[0],src.shape[1]
    
    for i in range(rows):
        for j in range(cols):
            dst[i,j] = src[rows - i - 1,j]
    
    return dst
```

```python
dst = mirror_y(src)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('RGB')
plt.subplot(122)
plt.imshow(dst)
plt.title('Mirror Y')
```
