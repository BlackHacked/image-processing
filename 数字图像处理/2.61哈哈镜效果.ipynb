# 哈哈镜效果

```python
#coding:utf-8
import numpy as np
%matplotlib  inline
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def distoring_mirror(src,x,y):
    dst = np.zeros(src.shape,dtype=np.float64)
    rows,cols = src.shape[0],src.shape[1]
    rgb = np.float64(src)
    raidu = 0.0
    theta = 0.0
    map_x = 0.0
    map_y = 0.0
    map_r = 0.0
    for i in range(rows):
        for j in range(cols):
            tx = j - x
            ty = i - y
            theta = np.arctan2(ty,tx)
            radius = np.sqrt(tx * tx + ty * ty)
            map_r = np.sqrt(radius * 100)
            map_x = x + int(map_r * np.cos(theta))
            map_y = y + int(map_r * np.sin(theta))
            
            dst[i,j] = src[map_y,map_x]
            
    return cv2.convertScaleAbs(dst)
            
```

```python
dst = distoring_mirror(src,200,200)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('DST')
```
