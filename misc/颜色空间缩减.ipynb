```python
#coding:utf-8
'''
颜色空间缩减
'''
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from pyexcel_xls import get_data
from cv2 import bioinspired
import cv2.bioinspired
```

```python
src = cv2.cvtColor(cv2.imread('datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
# 创建查找表
div = 64
table = []
for i in range(256):
    table.append((int(round(i / div * div))))
lut = np.array(table,dtype=np.uint8).reshape((1,256))

dst = cv2.LUT(src,lut)
```

```python
plt.imshow(src)
plt.title('SRC')
plt.xticks([]),plt.yticks([])
```

```python
plt.imshow(dst)
plt.title('LUT-DST')
plt.xticks([]),plt.yticks([])
```
