## 从 RGB 到 HSL 或 HSV 的转换

## HSL到RGB转换

首先，如果 s = 0，则结果的颜色是非彩色的、或灰色的。在这个特殊情况，r, g 和 b 都等于 l。注意 h 的值在这种情况下是未定义的。

```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from pyexcel_xls import get_data
import pandas as pd
from pandas import DataFrame
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def rgb2hsl(src):
    rgb = np.float32(src)
#     hsl = np.zeros(src.shape,dtype=np.float32)
    r = rgb[:,:,0]
    g = rgb[:,:,1]
    b = rgb[:,:,2]
    m,n = r.shape
    # 求L
    maxcolor = cv2.max(cv2.max(r,g),b)
    mincolor = cv2.min(cv2.min(r,g),b)
    L = (maxcolor + mincolor) / 2
    
    H = np.zeros_like(r)
    S = np.zeros_like(g)
    
    # 求S,H
    for i in range(m):
        for j in range(n):
            # 求S
            if maxcolor[i,j] == mincolor[i,j]:
                S[i,j] = 0
            else:
                if L[i,j] <= 0.5:
                    S[i,j] = (maxcolor[i,j] - mincolor[i,j]) / (2 * L[i,j])
                else:
                    S[i,j] = (maxcolor[i,j] - mincolor[i,j]) / (2 - 2 * L[i,j])
                    
            # 求H
            if maxcolor[i,j] == mincolor[i,j]:
                H[i,j] = 0
            elif r[i,j] == maxcolor[i,j]:
                if g[i,j] >= b[i,j]:
                    H[i,j] = 60 * (g[i,j] - b[i,j]) / (maxcolor[i,j] - mincolor[i,j])
                else:
                    H[i,j] = 60 * (g[i,j] - b[i,j]) / (maxcolor[i,j] - mincolor[i,j]) + 360
            elif g[i,j] == maxcolor[i,j]:
                H[i,j] = 120 + 60 * (b[i,j] - r[i,j]) / (maxcolor[i,j] - mincolor[i,j])
            else:
                H[i,j] = 240 + 60 * (r[i,j] - g[i,j]) / (maxcolor[i,j] - mincolor[i,j])
                
    dst = np.zeros_like(rgb)
    dst[:,:,0] = H
    dst[:,:,1] = S
    dst[:,:,2] = L
    return dst 
```

```python
def hsl2rgb(hsl):
    H = hsl[:,:,0]
    S = hsl[:,:,1]
    L = hsl[:,:,2]
    
    R = np.zeros_like(H)
    G = np.zeros_like(H)
    B = np.zeros_like(H)
    
    m,n = H.shape
    
    for i in range(m):
        for j in range(n):
            #求R
            if S[i,j] == 0:
                R[i,j] = L[i,j]
            elif L[i,j] <= 0.5:
                temp2 = L[i,j] * (1.0 + S[i,j])
            else:
                temp2 = L[i,j] + S[i,j] - L[i,j] * S[i,j]
            temp1 = 2 * L[i,j] - temp2
            h = H[i,j] / 360
            temp3 = h + 1 / 3
            if temp3 < 0:
                temp3 = temp3 + 1
            if temp3 > 1:
                temp3 = temp3 - 1
                
            if 6 * temp3 < 1:
                R[i,j] = temp2
            elif 2 * temp3 < 2:
                R[i,j] = temp1 + (temp2 - temp1) * ( 2 / 3 - temp3) * 6
            else:
                R[i,j] = temp1
            
            # 求G
            if S[i,j] == 0:
                G[i,j] = L[i,j]
            elif L[i,j] < 0.5:
                temp2 = L[i,j] * (1 + S[i,j])
            else:
                temp2 = L[i,j] + S[i,j] - L[i,j] * S[i,j]
                
            temp1 = 2 * L[i,j] - temp2
            h = H[i,j] / 360
            temp3 = h
            if temp3 < 0:
                temp3 = temp3 + 1
    
            if temp3 > 1:
                temp3 = temp3 - 1
            if 6 * temp3 < 1:
                G[i,j] = temp1 + (temp2 - temp1) * 6 * temp3
            elif 2 * temp3 < 1:
                G[i,j] = temp2
            elif 3 * temp3 < 2:
                G[i,j] = temp1 + (temp2 - temp1) * ( 2 / 3 - temp3) * 6
            else:
                G[i,j] = temp1
            
            # 求B
            if S[i,j] == 0:
                B[i,j] = L[i,j]
            elif L[i,j] <= 0.5:
                temp2 = L[i,j] * (1.0 + S[i,j])
            else:
                temp2 = L[i,j] + S[i,j] - L[i,j] * S[i,j]
            temp1 = 2 * L[i,j] - temp2
            h = H[i,j] / 360
            temp3 = h - 1 / 3
            if temp3 < 0:
                temp3 = temp3 + 1
            if temp3 > 1:
                temp3 = temp3 - 1
            
            if 6 * temp3 < 1:
                B[i,j] = temp1 + (temp2 - temp1) * 6 * temp3
            elif 2 * temp3 < 1:
                B[i,j] = temp1 + (temp2 - temp1) * (2 / 3 - temp3) * 6
            else:
                B[i,j] = temp1
                
    dst = np.zeros_like(hsl)
    dst[:,:,0] = R
    dst[:,:,1] = G
    dst[:,:,2] = B
    return dst
            
```

```python
def rgb2hsl_pixel(R,G,B):
     
    c_max = max(max(R,G),B)
    c_min = min(min(R,G),B)
    delta = c_max - c_min
    
    if delta == 0:
        H = 0
    elif c_max == R:
        H = 60 * (((G - B) / delta) % 6)
    elif c_max == G:
        H = 60 * (((B - R) / delta) + 2)
    elif c_max == B:
        H = 60 * (((R - G) / delta) + 4)
   
    H = round(H)
    L = round((c_max + c_min) / 2,2)
    
    if c_max == 0:
        S = 0
    else:
        S = round(delta / (1 - np.abs(2 * L - 1)), 2)
    return H,S,L
    
def rgb2hsl_2(src):
    rgb = np.float32(src) / 255
    dst = np.zeros_like(rgb)
    rows,cols = src.shape[0],src.shape[1]
    for i in range(rows):
        for j in range(cols):
            r = rgb[i,j,0]
            g = rgb[i,j,1]
            b = rgb[i,j,2]
            h,s,l = rgb2hsl_pixel(r,g,b)
            dst[i,j,0] = h
            dst[i,j,1] = s
            dst[i,j,2] = l
            
    return dst

def hsl2rgb_pixel(H,S,L):
    C = (1 - np.abs(2 * L - 1)) * S
    X = C * (1 - np.abs(((H / 60) % 2) - 1))
    m = L - (C / 2)
    
    if H >= 0 and H < 60:
        R = C
        G = X
        B = 0
    elif H >= 60 and H < 120:
        R = X
        G = C
        B = 0
    elif H >= 120 and H < 180:
        R = 0
        G = C
        B = X
    elif H>= 180 and H < 240:
        R = 0
        G = X
        B = C
    elif H >= 240 and H <300:
        R = X
        G = 0
        B = C
    else:
        R = C
        G = 0
        B = X
    
    R = np.ceil((R + m) * 255)
    G = np.ceil((G + m) * 255)
    B = np.ceil((B + m) * 255)
    
    return R,G,B

def hsl2rgb_2(hsl):
#     rgb = np.float32(src) / 255
    dst = np.zeros_like(hsl)
    rows,cols = src.shape[0],src.shape[1]
    for i in range(rows):
        for j in range(cols):
            h = hsl[i,j,0]
            s = hsl[i,j,1]
            l = hsl[i,j,2]
            r,g,b = hsl2rgb_pixel(h,s,l)
            dst[i,j,0] = r
            dst[i,j,1] = g
            dst[i,j,2] = b
            
    return dst
```

```python
def rgb2hsl_3(src):
    rgb = np.float32(src) / 255
    R = rgb[:,:,0]
    G = rgb[:,:,1]
    B = rgb[:,:,2]
    
    H = np.zeros_like(R)
    S = np.zeros_like(R)
    L = np.zeros_like(R)
    
    c_max = cv2.max(cv2.max(R,G),B)
    c_min = cv2.min(cv2.min(R,G),B)
    
    delta = (c_max - c_min) + (1.0e-5)
    df_delta = DataFrame(delta)
    idx = (df_delta == 0)
    H[idx] = 0
    
    delta_R = c_max - R
    df_R = DataFrame(delta_R)
    idx = (df_R == 0)
    H[idx] = 60 * (((G[idx] - B[idx]) / delta[idx]) % 6)
    
    delta_G = c_max - G
    df_G = DataFrame(delta_G)
    idx = (df_G == 0)
    H[idx] = 60 * (((B[idx] - R[idx]) / delta[idx]) + 2)
    
    delta_B = c_max - B
    df_B = DataFrame(delta_B)
    idx = (df_B == 0)
    H[idx] = 60 * (((R[idx] - G[idx]) / delta[idx]) + 4)
    
    H = np.round(H)
    L = np.round((c_max + c_min) / 2)
    
    df_max = DataFrame(c_max)
    idx = (df_max > 0)
    S[idx] = np.round(delta[idx] / (1 - np.abs(2 * L[idx] - 2)))
    
    dst = np.zeros_like(rgb)
    dst[:,:,0] = H
    dst[:,:,1] = S
    dst[:,:,2] = L
    
    return dst
        
def hsl2rgb_3(hsl):
    H = hsl[:,:,0]
    S = hsl[:,:,1]
    L = hsl[:,:,2]
    
    C = (1 - np.abs(2 * L - 2)) * S
    X = C * (1 - np.abs(((H / 60) % 2) - 1))
    m = L - (C / 2)
    
#     df_C = DataFrame(C)
#     df_X = DataFrame(X)
#     df_m = DataFrame(m)
    df_H = DataFrame(H)
    
    R = np.zeros_like(H)
    G = np.zeros_like(H)
    B = np.zeros_like(H)
#     C = np.zeros_like(H)
#     X = np.zeros_like(H)
#     m = np.zeros_like(H)
    
    idx = (df_H >= 0) & (df_H < 60)
#     C[idx] = (1 - np.abs(2 * L[idx] - 2)) * S[idx]
#     X[idx] = C[idx] * (1 - np.abs(((H[idx] / 60) % 2) - 1))
#     m[idx] = L[idx] - (C[idx] / 2)
    R[idx] = C[idx] + m[idx]
    G[idx] = X[idx] + m[idx]
    B[idx] = m[idx] 
    
    idx = (df_H >= 60) & (df_H < 120)
#     C[idx]= (1 - np.abs(2 * L[idx] - 2)) * S[idx]
#     X[idx] = C[idx] * (1 - np.abs(((H[idx] / 60) % 2) - 1))
#     m[idx] = L[idx] - (C[idx] / 2)
    R[idx] = X[idx] + m[idx]
    G[idx] = C[idx] + m[idx]
    B[idx] = m[idx]
    
    idx = (df_H >= 120) & (df_H < 180)
#     C[idx] = (1 - np.abs(2 * L[idx] - 2)) * S[idx]
#     X[idx] = C[idx] * (1 - np.abs(((H[idx] / 60) % 2) - 1))
#     m[idx] = L[idx] - (C[idx] / 2)
    R[idx] = m[idx]
    G[idx] = C[idx] + m[idx]
    B[idx] = X[idx] + m[idx]
    
    
    idx = (df_H >= 180) & (df_H < 240)
#     C[idx] = (1 - np.abs(2 * L[idx] - 2)) * S[idx]
#     X[idx] = C[idx] * (1 - np.abs(((H[idx] / 60) % 2) - 1))
#     m[idx] = L[idx] - (C[idx] / 2)
    R[idx] = m[idx]
    G[idx] = X[idx] + m[idx]
    B[idx] = C[idx] + m[idx]
    
    idx = (df_H >= 240) & (df_H < 300)
#     C[idx] = (1 - np.abs(2 * L[idx] - 2)) * S[idx]
#     X[idx] = C[idx] * (1 - np.abs(((H[idx] / 60) % 2) - 1))
#     m[idx] = L[idx] - (C[idx] / 2)
    R[idx] = X[idx] + m[idx]
    G[idx] = m[idx]
    B[idx] = C[idx] + m[idx]
    
    idx = (df_H >= 300) & (df_H <= 360)
#     C[idx] = (1 - np.abs(2 * L[idx] - 2)) * S[idx]
#     X[idx] = C[idx] * (1 - np.abs(((H[idx] / 60) % 2) - 1))
#     m[idx] = L[idx] - (C[idx] / 2)
    R[idx] = C[idx] + m[idx]
    G[idx] = m[idx]
    B[idx] = X[idx] + m[idx]
    
    dst = np.zeros_like(hsl)
    
    
    
    dst[:,:,0] = R
    dst[:,:,1] = G
    dst[:,:,2] = B
    
    return np.ceil(dst * 255)
    
```

```python
def rgb2hsl_matrix(colors):
    shape = colors.shape
    rgb = colors.reshape((colors.shape[0] * colors.shape[1],colors.shape[2]))
    colors = np.array(rgb)
#     colors=np.array(colors)
#     print(colors.shape)
    var_Rs=colors[:,0]/255.0
    var_Gs=colors[:,1]/255.0
    var_Bs=colors[:,2]/255.0
    var_Min=np.amin(colors/255.0,axis=1) # min(Ri,Gi,Bi) in each row
    var_Max=np.amax(colors/255.0,axis=1)
    del_Max=var_Max-var_Min
    ls=(var_Max+var_Min)/2.0
    hs=0.0
    ss=0.0
    # When del_Max=0,it may exist the situation that var_Max+var_Min or 2.0-var_Max-var_Min = 0
    # and the program would report error "invalid value encountered in divide"
    # However, at the end it will set zero when del_Max=0 
    # so I ignore the error here
    # https://docs.scipy.org/doc/numpy/reference/generated/numpy.seterr.html
    np.seterr(divide='ignore', invalid='ignore')
    ss=np.where(ls<0.5,del_Max/(var_Max+var_Min),del_Max/(2.0-var_Max-var_Min))
    
    del_R=np.where(del_Max!=0.0,(((var_Max-var_Rs)/6.0)+(del_Max/2.0))/del_Max,0.0)
    del_G=np.where(del_Max!=0.0,(((var_Max-var_Gs)/6.0)+(del_Max/2.0))/del_Max,0.0)
    del_B=np.where(del_Max!=0.0,(((var_Max-var_Bs)/6.0)+(del_Max/2.0))/del_Max,0.0)

    hs=np.where(var_Rs==var_Max,del_B-del_G,hs)
    hs=np.where(var_Gs==var_Max,(1.0/3.0)+del_R-del_B,hs)
    hs=np.where(var_Bs==var_Max,(2.0/3.0)+del_G-del_R,hs)

    hs=np.where(del_Max!=0,hs,0.0)
    hs=np.where(hs<0.0,hs+1.0,hs)
    hs=np.where(hs>1.0,hs-1.0,hs)
    ss=np.where(del_Max!=0,ss,0.0)

    hsl=np.vstack((hs,ss,ls)).transpose()
    hsl = np.reshape(hsl,shape)
    return hsl
```

```python
def hue2rgb_matrix(v1,v2,vH):
    vH=np.where(vH<0.0,vH+1.0,vH)
    vH=np.where(vH>1.0,vH-1.0,vH)
    res=np.where(vH>=2.0/3,v1,0.0)
    res=np.where(vH<2.0/3,v1 + (v2-v1)*((2.0/3.0)-vH)*6.0,res)
    res=np.where(vH<1.0/2,v2,res)
    res=np.where(vH<1.0/6,v1+(v2-v1)*6.0*vH,res)
    return res

def hsl2rgb_matrix(colors):
    shape = colors.shape
    hsl = np.reshape(colors,(shape[0] * shape[1],shape[2]))
    colors=np.array(hsl)
    hs=colors[:,0]
    ss=colors[:,1]
    ls=colors[:,2]
    Rs,Gs,Bs=(ls*255,)*3
    #np.seterr(divide='ignore', invalid='ignore')

    var_2=np.where(ls<0.5,ls*(1.0+ss),(ls+ss)-(ss*ls))
    var_1=2.0*ls-var_2
    Rs=255*hue2rgb_matrix(var_1,var_2,hs+(1.0/3.0))
    Gs=255*hue2rgb_matrix(var_1,var_2,hs)
    Bs=255*hue2rgb_matrix(var_1,var_2,hs-(1.0/3.0))
    Rs=np.where(ss==0.0,ls*255,Rs)
    Gs=np.where(ss==0.0,ls*255,Gs)
    Bs=np.where(ss==0.0,ls*255,Bs)
    Rs=np.maximum(np.minimum(Rs,255),0)
    Gs=np.maximum(np.minimum(Gs,255),0)
    Bs=np.maximum(np.minimum(Bs,255),0)
    RGBs=np.vstack((Rs,Gs,Bs)).transpose()  # 3*n -> n*3
    RGBs=RGBs.astype('int')
    RGBs = np.reshape(RGBs,shape)
    return RGBs
```
