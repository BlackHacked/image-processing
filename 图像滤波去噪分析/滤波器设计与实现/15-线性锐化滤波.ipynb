# 线性锐化滤波

```python
#coding:utf-8
%matplotlib inline
import numpy as np
import scipy
from scipy.special import comb
import matplotlib.pyplot as plt
import cv2
from scipy import stats
```

```python
src = cv2.imread('../../datas/f4.jpg',0)
rgb = cv2.cvtColor(cv2.imread('../../datas/f4.jpg'),cv2.COLOR_BGR2RGB)
```

```python
w = np.array([[-1 ,-1 ,-1],  
    [-1, 8, -1],
    [-1, -1, -1]])
```

```python
def linear_sharpen(img,w):
    img1 = cv2.filter2D(img,-1,w)
    dst = np.abs(img - img1)
    return dst
```

```python
dst = linear_sharpen(rgb,w)
```

```python
plt.imshow(dst)
```
