# 拉普拉斯滤波器

```python
#coding:utf-8
%matplotlib inline
import numpy as np
import scipy
from scipy.special import comb
import matplotlib.pyplot as plt
import cv2
from scipy import stats
```

```python
src = cv2.imread('../../datas/f4.jpg',0)
rgb = cv2.cvtColor(cv2.imread('../../datas/f4.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def laplacian_kernel(p = 0.2):
    alpha = p
    alpha = max(0,min(alpha,1))
    h1 = alpha / (alpha + 1)
    h2 = alpha / (alpha + 1)
    h = np.array([
        [h1,h2,h1],
        [h2 ,- 4 / (alpha + 1),h2],
        [h1,h2,h1]
    ])
    return h
```

```python
kernel = laplacian_kernel(0.5)
print(kernel)
```

```python
dst = cv2.filter2D(rgb.astype(np.float32),-1,kernel)
dst = np.abs(dst)
dst[dst > 255] = 255
dst = dst.astype(np.uint8)
```

```python
plt.imshow(dst)
```
