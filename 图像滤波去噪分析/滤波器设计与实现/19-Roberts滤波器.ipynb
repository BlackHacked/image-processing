# Roberts滤波器

```python
#coding:utf-8
%matplotlib inline
import numpy as np
import scipy
from scipy.special import comb
import matplotlib.pyplot as plt
import cv2
from scipy import stats
```

```python
src = cv2.imread('../../datas/f4.jpg',0)
rgb = cv2.cvtColor(cv2.imread('../../datas/f4.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def roberts(src):
    dst = np.zeros(src.shape,dtype=np.int32)
    src_i = src.astype(np.int32)
    rows,cols = src.shape[0],src.shape[1]
    for i in range(1,rows - 1):
        for j in range(1,cols - 1):
            dst[i,j] = np.abs(src_i[i + 1,j + 1] - src_i[i,j]) + \
            np.abs(src_i[i + 1,j] - src_i[i,j + 1])
    dst[dst > 255] = 255
    dst = dst.astype(np.uint8)
    return dst
```

```python
dst = roberts(rgb)
```
