# 图像保边滤波算法-Surface Blur(表面模糊滤波)算法与实现



```python
from scipy import signal
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
src = cv2.imread('../../datas/face3.jpg')
gray = cv2.imread('../../datas/face3.jpg',0)
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
def clip3(x,a,b):
    return min(max(a,x),b)

def surface_blur(img,radius=10,thresh=15):
    src = np.float32(img)
    dst = np.zeros_like(src)
    rows,cols = src.shape[0],src.shape[1]
    for i in range(rows):
        for j in range(cols):
            sumr = sumrw = sumg = sumgw = sumb = sumbw = 0
            for n in range(-radius,radius+1):
                for m in range(-radius,radius+1):
                    x = clip3(j+m,0,cols-1)
                    y = clip3(i+n,0,rows-1)
                    k = 1.0 - np.abs(src[y,x,0] - src[i,j,0]) / (2.5 * thresh)
                    sumr += k * src[y,x,0]
                    sumrw += k
                    k = 1.0 - np.abs(src[y,x,1] - src[i,j,1]) / (2.5 * thresh)
                    sumg += k * src[y,x,1]
                    sumgw += k
                    k = 1.0 - np.abs(src[y,x,0] - src[i,j,2]) / (2.5 * thresh)
                    sumb += k * src[y,x,2]
                    sumbw += k
            dst[i,j] = (sumr / sumrw,sumg / sumgw,sumb / sumbw)
    dst = np.clip(dst,0,255)
    dst = np.uint8(dst)
    return dst
```

```python
dst = surface_blur(src)
```
