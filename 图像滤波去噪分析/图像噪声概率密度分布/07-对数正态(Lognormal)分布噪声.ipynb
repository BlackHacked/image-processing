# 对数正态(Lognormal)分布噪声

```python
#coding:utf-8
%matplotlib inline
import numpy as np
import scipy
from scipy.special import comb
import matplotlib.pyplot as plt
import cv2
from scipy import stats
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def noise_lognormal(M,N,a = 1,b=0.5):
    arr = np.abs(np.random.randn(M,N) )+ 1.0e-7
    print(arr)
    x = np.log(arr) 
#     print(x)
    R = a * np.exp(b * x)
    return R
```

```python
noise = noise_lognormal(src.shape[0],src.shape[1],a=2,b=30)
```

```python
noise_uint = cv2.convertScaleAbs(noise * 255)
src_noise = np.zeros(src.shape,dtype=np.float64)
src_noise[:,:,0] = np.float64(src[:,:,0]) + noise * 255
src_noise[:,:,1] = np.float64(src[:,:,1]) + noise * 255
src_noise[:,:,2] = np.float64(src[:,:,2]) + noise * 255
```

```python
plt.imshow(noise_uint,'gray')
```
