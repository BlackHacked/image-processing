# 二项分布噪声

```python
#coding:utf-8
%matplotlib inline
import numpy as np
import scipy
from scipy.special import comb
import matplotlib.pyplot as plt
import cv2
from scipy import stats
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def noise_binom(im,M,N,b=0.5):
    R = np.zeros(im.shape,dtype=np.float64)
    a = np.floor(np.float64(im) / 30) + 1
    R = comb(a,np.floor(a / 2)) * \
    np.power(b,np.floor(a / 2)) * \
    np.power((1-b),a - np.floor(a / 2))
    return R
```

```python
noise = noise_binom(src,src.shape[0],src.shape[1])
# noise = noise_BB(src)
print(noise[:,:,0])
```

```python
print(cv2.convertScaleAbs(noise*255)[:,:,0])
```

```python
plt.imshow(cv2.convertScaleAbs(noise * 255))
```

```python
dst = np.float64(src) + (noise*255)
```

```python
plt.imshow(cv2.convertScaleAbs(dst))
```
