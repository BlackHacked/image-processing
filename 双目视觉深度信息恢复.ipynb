```python
#coding:utf-8
'''
双目视觉深度信息恢复
'''
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

src_left = cv2.imread('datas/left.png',0)
src_right = cv2.imread('datas/right.png',0)
```

```python
stereo = cv2.StereoSGBM_create(numDisparities=16, blockSize=15)
disparity = stereo.compute(src_left, src_right)  
  
plt.subplot(131),plt.imshow(src_left,'gray'),plt.title('img_left'),plt.xticks([]),plt.yticks([])  
plt.subplot(132),plt.imshow(src_right,'gray'),plt.title('right'),plt.xticks([]),plt.yticks([])  
plt.subplot(133),plt.imshow(disparity,'gray'),plt.title('disparity'),plt.xticks([]),plt.yticks([])  
plt.show()  
```
