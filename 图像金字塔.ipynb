```python
#coding:utf-8
'''
图像金字塔
'''
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from pyexcel_xls import get_data
```

```python
src = cv2.imread('datas/f2.jpg',0)
```

```python
# 先高斯滤波
gaussianblured = cv2.GaussianBlur(src,(3,3),0.5)
print(gaussianblured.shape)
# 再降采样
rows,cols = src.shape
y_idx = np.arange(0,rows,2).reshape((rows // 2,1))
x_idx = np.arange(0,cols,2).reshape((1,cols // 2))

sample = gaussianblured[(y_idx,x_idx)] #.reshape((rows // 2,cols // 2))
```

```python
print(sample.shape)
```

```python
plt.subplot(121)
plt.imshow(src,'gray')
plt.title('SRC')
plt.subplot(122)
plt.title('down sample')
plt.imshow(sample,'gray')
```

```python
# 使用OpenCV实现
dst = cv2.pyrDown(src)
plt.subplot(121)
plt.imshow(src,'gray')
plt.title('SRC')
plt.subplot(122)
plt.title('pyrDown')
plt.imshow(dst,'gray')
```

```python
# 向上采样
dst = cv2.pyrUp(src)
plt.subplot(121)
plt.imshow(src,'gray')
plt.title('SRC')
plt.subplot(122)
plt.title('pyrDown')
plt.imshow(dst,'gray')
```
