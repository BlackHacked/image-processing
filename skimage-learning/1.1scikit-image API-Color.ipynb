# scikit-image API-Color

```python
from skimage import data
import skimage.color as color
import numpy as np
%matplotlib  inline
import matplotlib.pyplot as plt
```

## 1.转换颜色空间

scikit-image支持的颜色空间有：‘RGB’, ‘HSV’, ‘RGB CIE’, ‘XYZ’, ‘YUV’, ‘YIQ’, ‘YPbPr’, ‘YCbCr’, ‘YDbDr’。可以使用skimage.color.convert_colorspace(arr,fromspace,tospace)方法

```python
img = data.astronaut()
img_hsv = color.convert_colorspace(img,'RGB','HSV')
plt.subplot(121)
plt.imshow(img)
plt.title('RGB')
plt.subplot(122)
plt.imshow(img_hsv)
plt.title('RGB->HSV')
```

## 2.猜测图像的维度

```python
dims = color.guess_spatial_dimensions(img)
print('dims:',dims)
```

## 3.更加方便的转换函数

### 3.1RGB与HSV相互转换

```python
img_hsv = color.rgb2hsv(img)
rgb = color.hsv2rgb(img_hsv)
plt.subplot(131)
plt.imshow(img)
plt.title('RGB')
plt.subplot(132)
plt.imshow(img_hsv)
plt.title('RGB->HSV')
plt.subplot(133)
plt.imshow(rgb)
plt.title('HSV->RGB')
```

### 3.2 RGB与XYZ相互转换

```python
img_xyz = color.rgb2xyz(img)
rgb = color.xyz2rgb(img_xyz)
plt.subplot(131)
plt.imshow(img)
plt.title('RGB')
plt.subplot(132)
plt.imshow(np.uint8(img_xyz * 255))
# plt.imshow(img_xyz)
plt.title('RGB->XYZ')
plt.subplot(133)
plt.imshow(rgb)
plt.title('XYZ->RGB')
```

### 3.3RGB与RGB CIE相互转换

```python
img_rgbcie = color.rgb2rgbcie(img)
rgb = color.rgbcie2rgb(img_rgbcie)
plt.subplot(131)
plt.imshow(img)
plt.title('RGB')
plt.subplot(132)
plt.imshow(np.uint8(np.abs(img_rgbcie * 255)))
# plt.imshow(img_rgbcie)
plt.title('RGB->RGB CIE')
plt.subplot(133)
plt.imshow(np.uint8(np.abs(rgb * 255)))
plt.title('RGB CIE->RGB')
```

### 3.4RGB转换灰度图像

使用转换公式：Y = 0.2125 R + 0.7154 G + 0.0721 B

```python
img_gray = color.rgb2grey(img)
rgb = color.rgbcie2rgb(img_rgbcie)
plt.subplot(121)
plt.imshow(img)
plt.title('RGB')
plt.subplot(122)
# plt.imshow(np.uint8(img_rgbcie * 255))
plt.imshow(img_gray,'gray')
plt.title('RGB->Grey')
```

```python
img_gray = color.rgb2gray(img)
rgb = color.rgbcie2rgb(img_rgbcie)
plt.subplot(121)
plt.imshow(img)
plt.title('RGB')
plt.subplot(122)
# plt.imshow(np.uint8(img_rgbcie * 255))
plt.imshow(img_gray,'gray')
plt.title('RGB->Gray')
```

### 3.5XYZ与LAB相互转换

```python
img_xyz = color.rgb2xyz(img)
img_lab = color.xyz2lab(img_xyz)
img_xyz = color.lab2xyz(img_lab)
plt.subplot(131)
plt.imshow(np.uint8(np.abs((img_xyz * 255))))
plt.title('XYZ')
plt.subplot(132)
# plt.imshow(np.uint8(img_rgbcie * 255))
plt.imshow(np.uint8(np.abs((img_lab))))
plt.title('XYZ->LAB')
plt.subplot(133)
plt.imshow(np.uint8(np.abs((img_xyz * 255))))
plt.title('LAB->XYZ')
```

### 3.6RGB与LAB相互转换

```python
img_lab = color.rgb2lab(img)
img_rgb = color.lab2rgb(img_lab)
plt.subplot(131)
plt.imshow(img)
plt.title('RGB')
plt.subplot(132)
plt.imshow(np.uint8(np.abs((img_lab))))
plt.title('RGB->LAB')
plt.subplot(133)
plt.imshow(img_rgb)
plt.title('LAB->RGB')
```
