# scikit-image API-draw

```python
from skimage import data
import skimage.color as color
import skimage.draw as draw
import numpy as np
%matplotlib  inline
import matplotlib.pyplot as plt
```

## 1.绘制直线

skimage的draw模块提供了一些基本绘图函数，比如直线、圆等。draw模块的绘制图元函数与opencv不同的是，draw模块绘图函数只返回坐标，需要进一步设置返回坐标位置的像素颜色。

```python
# 绘制坐标(0,0)到坐(511,511)的直线
rr,cc = draw.line(0,0,511,511)
img = np.zeros((512,512,3),dtype=np.uint8)
# 设置直线颜色
img[rr,cc] = [255,0,0]
rr,cc = draw.line(511,0,0,511)
img[rr,cc] = [0,255,0]
plt.imshow(img)
plt.title('Draw Line')
```

```python
# 绘制反锯齿直线
rr,cc,val = draw.line_aa(0,0,511,511)
img = np.zeros((512,512,3),dtype=np.uint8)
# 设置直线颜色
r = img[:,:,0]
g = img[:,:,1]
b = img[:,:,2]
r[rr,cc] = val * 255
g[rr,cc] = 0
b[rr,cc] = 0

rr,cc,val = draw.line_aa(511,0,0,511)
g[rr,cc] = val *  255
r[rr,cc] = 0
b[rr,cc] = 0
plt.imshow(img)
plt.title('Draw AA Line')
```

## 2.Bezier曲线

```python
img = np.zeros((512,512,3),dtype=np.uint8)
rr,cc = draw.bezier_curve(10,400,400,0,80,80,20)
img[rr,cc] = [0,255,0]
plt.imshow(img)
plt.title('Bezier Curve')
```

## 3.多边形

```python
img = np.zeros((512,512,3),dtype=np.uint8)
r = np.array([10,200,80,100])
c = np.array([10,70,40,100])

rr,cc = draw.polygon(r,c)
img[rr,cc] = [255,0,0]
plt.imshow(img)
plt.title('Polygon')
```

多边形周长

```python
img = np.zeros((512,512,3),dtype=np.uint8)
rr,cc = draw.polygon_perimeter([50,-10,50,10],[-10,50,110,50],shape=img.shape,clip=True)
img[rr,cc] = [255,0,0]
plt.imshow(img)
plt.title('Polygon Perimeter')
```

## 4.绘制椭圆

```python
img = np.zeros((512,512,3),dtype=np.uint8)
# 中心坐标(256,256) 半径分别为200，60
rr,cc = draw.ellipse(256,256,200,60)
img[rr,cc] = [255,0,0]
# 旋转45度
rr,cc = draw.ellipse(256,256,200,60,rotation=np.deg2rad(45))
img[rr,cc] = [0,255,0]
plt.imshow(img)
plt.title('Ecllipse')
```

## 4.绘制圆

```python
img = np.zeros((512,512,3),dtype=np.uint8)
# 中心坐标(256,256) 半径分别为200
rr,cc = draw.circle(256,256,200)
img[rr,cc] = [255,0,0]
plt.imshow(img)
plt.title('Circle')
```

## 5.绘制矩形

```python
img = np.zeros((512,512,3),dtype=np.uint8)
# 中心坐标(256,256) 半径分别为200
rr,cc = draw.rectangle((10,10),extent=(210,210))
img[rr,cc] = [255,0,0]
plt.imshow(img)
plt.title('Circle')
```
