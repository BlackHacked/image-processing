```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import skimage as sk
import skimage.io as io
import skimage.color as color
import cv2
import scipy
import scipy.misc as sc
```

```python
src = io.imread('../datas/f4.jpg')
```

## 1.模糊

```python
dst = sc.imfilter(src,'blur')
plt.imshow(dst)
plt.title('Blur')
```

## 2.轮廓检测

```python
dst = sc.imfilter(src,'contour')
plt.imshow(dst)
plt.title('Contour')
```

## 3.细节强化

```python
dst = sc.imfilter(src,'detail')
plt.imshow(dst)
plt.title('Detail')
```

## 4.边缘增强

```python
dst = sc.imfilter(src,'edge_enhance')
plt.imshow(dst)
plt.title('Edge Enhance')
```

```python
dst = sc.imfilter(src,'edge_enhance_more')
plt.imshow(dst)
plt.title('Edge Enhance')
```

## 5.浮雕

```python
dst = sc.imfilter(src,'emboss')
plt.imshow(dst)
plt.title('Emboss')
```

## 6.边缘检测

```python
dst = sc.imfilter(src,'find_edges')
plt.imshow(dst)
plt.title('Find Edges')
```

## 7.平滑

```python
dst = sc.imfilter(src,'smooth')
plt.imshow(dst)
plt.title('Smooth')
```

```python
dst = sc.imfilter(src,'smooth_more')
plt.imshow(dst)
plt.title('Smooth')
```

## 8.图像锐化

```python
dst = sc.imfilter(src,'sharpen')
plt.imshow(dst)
plt.title('Sharpen')
```
