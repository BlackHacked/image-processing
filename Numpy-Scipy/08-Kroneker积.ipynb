```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
src1 = cv2.imread('../datas/f1.jpg',0)
src2 = cv2.imread('../datas/f2.jpg',0)
```

```python
rows,cols = src1.shape
src1 = cv2.resize(src1,(cols // 4,rows // 4)).astype(np.float32) / 255.0
src2 = cv2.resize(src2,(cols // 4,rows // 4)).astype(np.float32) / 255.0
```

```python
dst = np.kron(src1,src2)
```

```python
plt.imshow(dst,'gray')
```
