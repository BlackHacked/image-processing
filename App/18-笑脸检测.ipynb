# 笑脸检测

```python
import numpy as np
%matplotlib  inline
import cv2
from scipy import ndimage
import matplotlib.pyplot as plt
```

```python
f1 = cv2.cvtColor(cv2.imread('../datas/face6.jpg'),cv2.COLOR_BGR2RGB)
f1_gray = cv2.imread('../datas/face6.jpg',0)
f2 = cv2.cvtColor(cv2.imread('../datas/face5.jpg'),cv2.COLOR_BGR2RGB)
f2_gray = cv2.imread('../datas/face5.jpg',0)
```

```python
face_csc = cv2.CascadeClassifier('../datas/haarcascade_smile.xml')
```

```python
faces = face_csc.detectMultiScale(f2_gray)
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(f2, (x,y), (x+w, y+h), (0,255,0), 2)
```
