# 过度曝光原理

过度曝光原理：图像翻转，然后求原图与反图的最小值

```python
import numpy as np
import cv2
import random
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
src =  cv2.imread("../datas/exposure.jpg");
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```
