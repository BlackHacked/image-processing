# LSD直线检测

LSD是一种直线检测分割算法，它能在线性的时间内得出亚像素级精度的检测结果。该算法被设计成可以在任何数字图像上都无需参数调节。它可以自己控制误检的数量：平均而言，每张图有一个误检。

相关知识介绍

LSD的目标在于检测图像中局部的直的轮廓。这也是我们称之为直线分割的原因。轮廓是图像中的某些特殊区域，在这些区域，图像的灰度从黑到白或者从白到黑的剧烈变化。因此，梯度和level-line是两个重要的感念，如下图所示：

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
import skimage
import skimage.io
import imageio
import math
```

```python
src = imageio.imread('../datas/desktop.jpg')
gray = cv2.cvtColor(src,cv2.COLOR_RGB2GRAY)
```

```python
edge = cv2.Canny(gray,50,200,1)
```

```python
plt.imshow(edge,'gray')
```

```python
lsd = cv2.createLineSegmentDetector(cv2.LSD_REFINE_STD)
```

```python
lines,width,prec,nfa = lsd.detect(edge)
```

```python
dst = lsd.drawSegments(src,lines)
```

```python
plt.imshow(dst)
```
