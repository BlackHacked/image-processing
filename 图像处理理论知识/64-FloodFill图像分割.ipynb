```python
import numpy as np
%matplotlib  inline
import cv2
from scipy import ndimage
import matplotlib.pyplot as plt
```

```python
# read image
filename = '../datas/scenery.jpg'
im = cv2.imread(filename)
h,w = im.shape[:2]
# flood fill example
diff = (6,6,6)
mask = np.zeros((h+2,w+2),np.uint8)
dst = cv2.floodFill(im,mask,(10,10), (255,255,0),diff,diff)
```
