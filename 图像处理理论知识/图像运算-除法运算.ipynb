# 图像运算-除法运算

除法运算可以用于校正成像设备的非线性影响，还可以用于检测两幅图像间的区别。除法运算的结果是像素值的变化比率，而不是每个像素的绝对差异，因此，除法运算也称为比率变换。计算公式如下：

$$f(Z) = \frac{f(X)}{f(Y)}$$

其中，$f(Z)$是输出图像，$f(X)$，$f(Y)$是输入图像。在计算过程中还需要考虑像素值为零的情况。
