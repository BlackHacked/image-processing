# 图像运算-图像减法

图像减法也称为差分方法，是一种常用于检测图像变化以及运动物体的图像处理方法。可以使用以下公式描述：

$$f(D) = f(A) - f(B)$$

其中，$f(D)$是输出图像，$f(A)$和$f(B)$是输入图像，要求这两个图像必须有相同的维度。同时也要考虑两个输入图像相减后的负值处理，一般情况下，把负值设置为0。

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import skimage as sk
import skimage.io as io
import skimage.color as color
```

```python
A = io.imread('../datas/f2.jpg')
grayA = color.rgb2gray(A)
B = io.imread('../datas/f1.jpg')
grayB = color.rgb2gray(B)
```

```python
def imsubtract(A,B):
    dst = A - B
    dst = np.clip(dst,0,255)
    return dst
```

```python
dst = imsubtract(grayA,grayB)
plt.imshow(dst,'gray')
```
