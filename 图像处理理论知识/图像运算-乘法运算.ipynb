# 图像运算-乘法运算

两幅维度的图像乘法运算可以实现掩模操作，即屏蔽图像的某些部分。计算公式描述如下：

$$f(D) = f(A)f(B)$$

其中，$f(D)$为输出图像，$f(A)$、$f(B)$为输入图像。

如果一幅图像乘以一个常数，通常称为为缩放。如果这个常量小于1，则图像会变暗；如果大于1，则图像会变亮。可以使用以下公式描述 ：

$$f(D) = \alpha*f(A)$$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import skimage as sk
import skimage.io as io
import skimage.color as color
```

```python
A = io.imread('../datas/f2.jpg')
grayA = color.rgb2gray(A)
B = io.imread('../datas/f1.jpg')
grayB = color.rgb2gray(B)
```

```python
def immultiply(A,B):
    Af = np.float32(A)
    Bf = np.float32(B)
    dst = np.zeros_like(Af)
    dst = Af * Bf
    dst = np.clip(dst,0,255)
    return np.uint8(dst)
```

```python
def immultiply_constant(A,c):
    Af = np.float32(A)
    dst = np.zeros_like(Af)
    dst = c * Af
    dst = np.clip(dst,0,255)
    return np.uint8(dst)
```

```python
dst = immultiply(A,B)
plt.imshow(dst)
```

```python
dst = immultiply_constant(A,1.1)
plt.imshow(dst)
```

```python
import cv2
dst = cv2.multiply(A,B)
plt.imshow(dst)
```
