```python
#coding:utf-8
'''
图像噪声
https://blog.csdn.net/zhoufan900428/article/details/37695357
'''
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
src = cv2.cvtColor(cv2.imread('datas/f2.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_RGB2GRAY)
```

```python
def gaussian_noise(a,b,M,N):
    noise = a + b * np.random.randn(M,N)
    return noise
```

```python
M = gray.shape[0]
N = gray.shape[1]
a = 0
b = 0.08
noise = gaussian_noise(a,b,M,N)
print(noise.dtype)
plt.imshow(noise,cmap='gray')
```

```python
# dst = cv2.convertScaleAbs(cv2.add(noise,np.float64(gray)))
# plt.imshow(dst,cmap='gray')
dst = np.float64(gray) + noise
dst = cv2.convertScaleAbs(dst)
plt.imshow(dst,cmap='gray')
```
