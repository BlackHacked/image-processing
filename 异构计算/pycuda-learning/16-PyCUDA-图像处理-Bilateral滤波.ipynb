# PyCUDA-图像处理-Bilateral滤波

```python
import pycuda.autoinit
import pycuda.driver as cu
from pycuda import gpuarray
from pycuda.reduction import ReductionKernel
from pycuda.elementwise import ElementwiseKernel
import time
import numpy as np
from pycuda.compiler import SourceModule
import string
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import  animation
```

```python
def load_kernel_from_file(fname,kname):
    sources = None
    kernel = None
    with open(fname) as f:
        sources = f.read() # 通过str函数转换为unicode编码
        sm = SourceModule(str(sources))
        kernel = sm.get_function(kname)
    return kernel

def load_kernel_from_string(sources,kname):
    sm = SourceModule(str(sources))
    kernel = sm.get_function(kname)
    return kernel
```

```python
bilateral_filter_2d = load_kernel_from_file('../../datas/cuda/23-bilateral.cu','bilateral_filter_2d')
```

```python
src = cv2.cvtColor(cv2.imread('../../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_RGB2GRAY)
```

```python
rows,cols = gray.shape
block = (16,16,1)
grid = (int(cols / block[0]),int(rows / block[1]),1)
```

```python
dst = np.zeros_like(gray)
```

```python
bilateral_filter_2d(cu.In(gray),cu.Out(dst),np.int32(cols),np.int32(rows),grid=grid,block=block)
```

```python
plt.subplot(121)
plt.imshow(gray,'gray')
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst,'gray')
plt.title('Bilateral Filter')
```

```python
plt.imshow(np.abs(dst - gray),'gray')
plt.title('Different')
```
