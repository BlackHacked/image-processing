# 矩阵特征值分解

```python
import numpy as np
import scipy
import scipy.linalg
import skimage
import skimage.io
import skimage.color
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
img = skimage.io.imread('../datas/f2.jpg')
gray = skimage.color.rgb2gray(img)
```

```python
plt.imshow(gray,'gray')
```

```python
U,s,Vh = scipy.linalg.svd(gray)
```

```python
print(Vh.shape)
```

```python
plt.figure()
plt.imshow(U,'gray')
plt.figure()
plt.imshow(Vh,'gray')
```

```python
A = np.dot(U[:,0:32],np.dot(np.diag(s[0:32]),Vh[0:32,:]))
```

```python
plt.figure()
plt.imshow(gray,'gray')
plt.figure()
plt.imshow(A,'gray')
```
