```python
#coding:utf-8
'''
Numpy数组位操作
https://www.yiibai.com/numpy/numpy_binary_operators.html
'''
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
r = src[:,:,0]
g = src[:,:,1]
b = src[:,:,2]
```

```python
# 右移操作
dst_r = r.copy() >> 6
dst_r_2 = np.right_shift(r.copy(),2)
plt.subplot(131)
plt.imshow(r,'gray')
plt.title('SRC')
plt.subplot(132)
plt.imshow(dst_r,'gray')
plt.title('DST')
plt.subplot(133)
plt.imshow(dst_r_2,'gray')
plt.title('DST_2')
```

```python
# 左移操作
dst = src << 2
dst_2 = np.left_shift(src,4)
plt.subplot(131)
plt.imshow(src)
plt.title('SRC')
plt.subplot(132)
plt.imshow(dst)
plt.title('DST')
plt.subplot(133)
plt.imshow(dst_2)
plt.title('DST_2')
```
