# 几何分布噪声

```python
#coding:utf-8
%matplotlib inline
import numpy as np
import scipy
from scipy.special import comb
import matplotlib.pyplot as plt
import cv2
from scipy import stats
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def noise_geometry(im,b = 0.5):
    R = np.zeros(im.shape,dtype=np.float64)
    a = np.floor(im / 30) + 1
    R = b * (1 - b) ** (a - 1)
    return R
```

```python
noise = noise_geometry(src,0.25)
```

```python
noise_uint = cv2.convertScaleAbs(noise * 255)
src_noise = np.zeros(src.shape,dtype=np.float64)
src_noise = np.float64(src) + noise * 255
# src_noise[:,:,0] = np.float64(src[:,:,0]) + noise * 255
# src_noise[:,:,1] = np.float64(src[:,:,1]) + noise * 255
# src_noise[:,:,2] = np.float64(src[:,:,2]) + noise * 255
```
