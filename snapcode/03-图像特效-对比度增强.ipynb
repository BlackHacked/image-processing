# 图像特效-对比度增强

```python
#coding:utf-8
%matplotlib  inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
src = cv2.cvtColor(cv2.imread('../datas/dark.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.imread('../datas/f2.jpg',0)
```

```python
def adjust_contrast(src,degree):
    srcf = src.astype(np.float32)
    dst = np.zeros_like(srcf)
    if degree < -100:
        degree = -100
    if degree > 100:
        degree = 100
    contrast = (100 + degree) / 100.0
    contrast *= contrast
    dst = (srcf / 255.0) * contrast * 255.0
    dst[dst < 0] = 0
    dst[dst > 255] = 255
    return dst.astype(np.uint8)
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('SRC')
plt.subplot(122)
dst = adjust_contrast(src,50)
plt.imshow(dst)
plt.title('Adjust Contrast')
```
