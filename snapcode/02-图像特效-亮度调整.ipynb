# 图像特效-亮度调整

```python
#coding:utf-8
%matplotlib  inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
src = cv2.cvtColor(cv2.imread('../datas/dark.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.imread('../datas/f2.jpg',0)
```

```python
def brightness(src,cb):
    if cb < -255:
        cb = -255
    if cb > 255:
        cb = 255
    srci = src.astype(np.int32)
    dst = np.zeros_like(srci)
    
    dst = src + cb
    return cv2.convertScaleAbs(dst,20)
```

```python
dst = brightness(src,20)
plt.imshow(dst)
```
