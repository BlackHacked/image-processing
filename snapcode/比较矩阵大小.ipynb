# 比较矩阵大小

```python
#coding:utf-8
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
R = src[:,:,0]
G = src[:,:,1]
B = src[:,:,2]

r_rows_max = np.amax(R,axis = 1)
r_cols_max = np.amax(R,axis = 0)
g_max = np.amax(G,axis=1)
b_max = np.amax(B,axis=1)
print(r_cols_max.shape)
```

```python
max_val = cv2.max(cv2.max(R,G),B)
```

```python
print(max_val)
```

```python
plt.imshow(max_val,'gray')
```

```python
min_val = cv2.min(cv2.min(R,G),B)
plt.imshow(min_val,'gray')
```

```python
diff = max_val - min_val
plt.imshow(diff,'gray')
```

```python
edge = cv2.Canny(diff,180,255)
plt.imshow(edge,'gray')
```
