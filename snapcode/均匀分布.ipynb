# 均匀分布

```python
#coding:utf-8
import numpy as np
import cv2
import scipy
from scipy import linalg
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def noise_uniform(M,N,a = 0,b = 1):
    R = a + (b - a) * np.random.randn(M,N)
    return R
```

```python
R = np.float32(src[:,:,0].copy()) 
G = np.float32(src[:,:,1].copy()) 
B = np.float32(src[:,:,2].copy()) 

noise = noise_uniform(R.shape[0],R.shape[1],50,60)
R = R + noise
G = G + noise
B = B + noise
dst = np.zeros(src.shape,dtype=np.float32)
dst[:,:,0] = R
dst[:,:,1] = G
dst[:,:,2] = B
```

```python
plt.imshow(noise,'gray')
```

```python
plt.imshow(cv2.convertScaleAbs(dst))
```

```python
plt.imshow((cv2.convertScaleAbs(dst) - src)[:,:,2],'gray')
```
