# 显示中文

```python
%matplotlib inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image,ImageDraw,ImageFont
```

```python
src = cv2.imread('../datas/f4.jpg')
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
pil_im = Image.fromarray(rgb)
```

```python
draw = ImageDraw.Draw(pil_im)
```

```python
font = ImageFont.truetype("../datas/fonts/msyh.ttf", 40, encoding="utf-8")
```

```python
draw.text((0, 0), "Hello，中文", (255, 0, 0), font=font) 
```

```python
dst = np.array(pil_im)
```

```python
plt.imshow(dst)
```
