# 指数分布噪声

```python
%matplotlib inline
import numpy as np
import scipy
from scipy.special import comb
import matplotlib.pyplot as plt
import cv2
from scipy import stats
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
```

```python
def noise_exponential(M,N,a=1,b=0.25):
    R = a * np.exp(b * np.random.randn(M,N))
    return R
```

```python
noise = noise_exponential(src.shape[0],src.shape[1],0.5,1.5)
```

```python
print(noise)
```

```python
noise_uint = cv2.convertScaleAbs(noise * 255)
print(noise_uint)
```

```python
src_noise = np.zeros(src.shape,dtype=np.float64)
src_noise[:,:,0] = np.float64(src[:,:,0]) + noise * 255
src_noise[:,:,1] = np.float64(src[:,:,1]) + noise * 255
src_noise[:,:,2] = np.float64(src[:,:,2]) + noise * 255
```

```python
plt.imshow(noise_uint,'gray')
```

```python
plt.imshow(src)
```

```python
plt.imshow(cv2.convertScaleAbs(src_noise))
```
