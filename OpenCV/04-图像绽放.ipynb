# 图像缩放

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
```

```python
src = cv2.imread('../datas/f4.jpg')
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
```

## 1.金字塔向上采样（图像放大）

```python
rows,cols = gray.shape
dst_pyrup = cv2.pyrUp(rgb,(cols * 2,rows * 2))
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst_pyrup)
plt.title('pyrUp')
```

## 2.金字塔向下采样（图像缩小）

```python
dst_pyrdown = cv2.pyrDown(rgb,(cols // 2 ,rows // 2))
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst_pyrdown)
plt.title('pyrDown')
```

## 3.使用resize函数

```python
dst = cv2.resize(rgb,(cols * 2,rows * 2))
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('resize:zoom out')
```

```python
dst = cv2.resize(rgb,(cols // 2,rows // 2))
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('resize:zoom in')
```
