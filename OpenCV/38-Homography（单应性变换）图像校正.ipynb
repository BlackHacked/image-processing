## Homography（单应性变换）图像校正

```python
import numpy as np
import cv2
import random
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
im1 =  cv2.imread("../datas/book1.jpg");
im2 =  cv2.imread("../datas/book2.jpg");
 
# Convert images to grayscale
im1_gray = cv2.cvtColor(im1,cv2.COLOR_BGR2GRAY)
im2_gray = cv2.cvtColor(im2,cv2.COLOR_BGR2GRAY)
```

```python
if __name__ == '__main__' :  
      
   
    # Read source image.  
    im_src = im2#cv2.imread('book2.jpg')  
    # Four corners of the book in source image  
    pts_src = np.array([[167.0, 264.0], [482.0, 798.0], [1079.0, 403.0],[613.0, 84.0]])  
   
   
    # Read destination image.  
    im_dst = im1#cv2.imread('book1.jpg')  
    # Four corners of the book in destination image.  
    pts_dst = np.array([[193.0, 742.0],[996.0, 874.0],[1059.0, 157.0],[266.0, 145.0]])  
   
    # Calculate Homography  
    h, status = cv2.findHomography(pts_src, pts_dst)  
       
    # Warp source image to destination based on homography  
    im_out = cv2.warpPerspective(im_src, h, (im_dst.shape[1],im_dst.shape[0]))  
       
    plt.figure(), plt.imshow(im_src[:, :, ::-1]), plt.title('src')
    plt.figure(), plt.imshow(im_dst[:, :, ::-1]), plt.title('dst')  
    plt.figure(), plt.imshow(im_out[:, :, ::-1]), plt.title('out')
```
