# 不规则ROI

```python
import numpy as np
import cv2
import random
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
src =  cv2.imread("../datas/sunset2.jpg")
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
gray = cv2.imread("../datas/sunset2.jpg",0)
```

```python
roi = np.zeros_like(rgb)
rows,cols = gray.shape
cx,cy = cols // 2,rows // 2
cv2.circle(roi,(cx,cy),rows // 2,(255,255,255),-1)
plt.imshow(roi)
```

```python
res = rgb & roi
plt.imshow(res)
```
