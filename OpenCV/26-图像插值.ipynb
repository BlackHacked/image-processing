# 图像插值

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei'] # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题
```

```python
src = cv2.imread('../datas/f2.jpg')
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
```

## 1.最近邻插值

```python
rows,cols = gray.shape
dst = cv2.resize(rgb,(int(cols * 1.5),int(rows * 1.5)),cv2.INTER_NEAREST)
plt.imshow(dst)
plt.title(u'最近邻插值')
```

## 2.双线性插值

```python
rows,cols = gray.shape
dst = cv2.resize(rgb,(int(cols * 1.5),int(rows * 1.5)),cv2.INTER_LINEAR)
plt.imshow(dst)
plt.title(u'双线性插值')
```

## 3.像素区域

```python
rows,cols = gray.shape
dst = cv2.resize(rgb,(int(cols * 1.5),int(rows * 1.5)),cv2.INTER_AREA)
plt.imshow(dst)
plt.title(u'像素区域')
```

## 4.三次插值

```python
rows,cols = gray.shape
dst = cv2.resize(rgb,(int(cols * 1.5),int(rows * 1.5)),cv2.INTER_CUBIC)
plt.imshow(dst)
plt.title(u'三次插值')
```
