# 艺术笔刷

```python
import numpy as np
import cv2
import random
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
src = cv2.imread('../datas/cow.jpg')
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
```

## Step 1:保边滤波器

```python
dst = cv2.edgePreservingFilter(rgb, flags=1, sigma_s=60, sigma_r=0.4)
```

```python
plt.imshow(dst)
```

## Step 2:细节增强

```python
dst = cv2.detailEnhance(dst, sigma_s=10, sigma_r=0.15)
```

```python
plt.imshow(dst)
```

## Step3:铅笔素描

```python
dst_gray, dst_color = cv2.pencilSketch(rgb, sigma_s=60, sigma_r=0.07, shade_factor=0.05)
```

```python
plt.imshow(dst_gray,'gray')
```

```python
plt.imshow(dst_color)
```

## Step 4:样式化

```python
dst = cv2.stylization(rgb, sigma_s=60, sigma_r=0.07)
```

```python
plt.imshow(dst)
```
