# 条码识别

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import imageio
import numpy as np
import pyzbar
from pyzbar.pyzbar import decode
```

```python
src = cv2.imread('../datas/zbar-test.jpg',0)
rgb = cv2.imread('../datas/zbar-test.jpg')
```

```python
results = decode(src)
```

```python
for result in results:
    print(result.data)
plt.imshow(rgb)
```

```python
rows,cols = src.shape
for result in results:
    x = result.rect.left
    y = result.rect.top
    w = result.rect.width
    h = result.rect.height
    roi = src[y:y+h,x:x+w]
    plt.figure()
    plt.imshow(roi,'gray')
```
