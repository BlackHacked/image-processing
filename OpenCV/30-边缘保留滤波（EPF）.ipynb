# 边缘保留滤波（EPF）

```python
from scipy import signal
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
src = cv2.imread('../datas/face2.jpg')
gray = cv2.imread('../datas/face2.jpg',0)
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
def bi(image):  
    """ 
    色彩窗的半径 
    图像将呈现类似于磨皮的效果 
    """  
  
    #image：输入图像，可以是Mat类型，  
    #       图像必须是8位或浮点型单通道、三通道的图像  
    #0：表示在过滤过程中每个像素邻域的直径范围，一般为0  
    #后面两个数字：空间高斯函数标准差，灰度值相似性标准差  
    dst=cv.bilateralFilter(image,0,60,10);  
    return dst
  
def shift(image):  
    """ 
    均值迁移 
    图像会呈现油画效果 
    """  
  
    #10:空间窗的半径  
    #50:色彩窗的半径  
    dst=cv2.pyrMeanShiftFiltering(image,10,50);
    return dst
```
