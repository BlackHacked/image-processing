# 曝光融合

```python
import cv2
import numpy as np
import sys
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def readImagesAndTimes():
    filenames = [
               "../datas/exposure-images/memorial0061.jpg",
               "../datas/exposure-images/memorial0062.jpg",
               "../datas/exposure-images/memorial0063.jpg",
               "../datas/exposure-images/memorial0064.jpg",
               "../datas/exposure-images/memorial0065.jpg",
               "../datas/exposure-images/memorial0066.jpg",
               "../datas/exposure-images/memorial0067.jpg",
               "../datas/exposure-images/memorial0068.jpg",
               "../datas/exposure-images/memorial0069.jpg",
               "../datas/exposure-images/memorial0070.jpg",
               "../datas/exposure-images/memorial0071.jpg",
               "../datas/exposure-images/memorial0072.jpg",
               "../datas/exposure-images/memorial0073.jpg",
               "../datas/exposure-images/memorial0074.jpg",
               "../datas/exposure-images/memorial0075.jpg",
               "../datas/exposure-images/memorial0076.jpg"
               ]
    images = []
    for filename in filenames:
        im = cv2.imread(filename)
        images.append(im)
  
    return images
```

```python
images = readImagesAndTimes()
needsAlignment = True
```

```python
# 对齐图像
alignMTB = cv2.createAlignMTB()
alignMTB.process(images, images)
```

```python
# 使用曝光就去融合图像
mergeMertens = cv2.createMergeMertens()
exposureFusion = mergeMertens.process(images)
```

```python
dst = exposureFusion * 255
dst = np.clip(dst,0,255)
dst = np.uint8(dst)
dst = cv2.cvtColor(dst,cv2.COLOR_BGR2RGB)
```
