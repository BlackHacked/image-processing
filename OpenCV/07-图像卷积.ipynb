# 图像卷积

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
src = cv2.imread('../datas/f4.jpg')
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
rgb_f = rgb.astype(np.float32)
```

```python
kernel = np.array([
    [0,-1,0],
    [-1,5,-1],
    [0,-1,-1]
])
dst = cv2.filter2D(rgb,-1,kernel)
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Image Convolution')
```

```python
kernel = np.array([
    [-1,-1,-1],
    [-1,8,-1],
    [-1,-1,-1]
])
dst = cv2.filter2D(rgb,-1,kernel)
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Image Convolution')
```

```python
kernel = np.ones((3,3),dtype=np.float32)
kernel[:] = -1
kernel[1,1] = 8.9
dst = cv2.filter2D(rgb_f,-1,kernel)
dst[dst < 0] = 0
dst[dst > 255] = 255
dst = dst.astype(np.uint8)
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Image Convolution')
```

```python
kernel = np.zeros((3,3),dtype=np.float32)
kernel[2,1] = - 1.
kernel[0,1] = -1.
kernel[1,1] = 5.
dst = cv2.filter2D(rgb_f,-1,kernel)
dst[dst < 0] = 0
dst[dst > 255] = 255
dst = dst.astype(np.uint8)
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Image Convolution')
```
