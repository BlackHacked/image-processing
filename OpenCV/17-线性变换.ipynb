# 线性变换

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
src = cv2.imread('../datas/f4.jpg')
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
rgb_f = rgb.astype(np.float32)
```

```python
def linear_transform(src,a,b):
    srcf = src.astype(np.float32)
    dst = a * src + b
    dst[dst < 0] = 0
    dst[dst > 255] = 255
    dst = dst.astype(np.uint8)
    return dst
```

```python
dst = linear_transform(rgb,1.2,-50)
```

```python
plt.imshow(dst)
```

```python
dst = linear_transform(rgb,-1,255)
plt.imshow(dst)
```
