# Gamma变换

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
src = cv2.imread('../datas/f4.jpg')
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
rgb_f = rgb.astype(np.float32)
```

```python
def gamma_transform(src,factor):
    lut = np.zeros((256,1))
    for i in range(256):
        lut[i] = ((i / 255.0) ** factor) * 255
    dst = cv2.LUT(src,lut)
    return dst
```

```python
dstr = gamma_transform(rgb[:,:,0],2.4)
dstg = gamma_transform(rgb[:,:,1],2.4)
dstb = gamma_transform(rgb[:,:,2],2.4)
dst = cv2.merge([dstr,dstg,dstb])
dst = dst.astype(np.uint8)
```
