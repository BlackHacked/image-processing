# 透明图像混合

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import imageio
import numpy as np
```

```python
img_bg = cv2.imread('../datas/backGround.jpg')
img_fg = cv2.imread('../datas/foreGroundAsset.png',-1)
```

```python
b,g,r,a = cv2.split(img_fg)
foreground = cv2.merge((b,g,r))
```

```python
alpha = cv2.merge((a,a,a))
```

```python
background = img_bg.astype(np.float32)
foreground = foreground.astype(np.float32)
alpha = alpha.astype(np.float32)/255.0
```

```python
# 混合
foreground = cv2.multiply(alpha, foreground)
background = cv2.multiply(1.0 - alpha, background)
outImage = cv2.add(foreground, background)
```

```python
outImage = cv2.cvtColor(outImage,cv2.COLOR_BGR2RGB)
outImage = np.clip(outImage,0,255)
outImage = np.uint8(outImage)
```
