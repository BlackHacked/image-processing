# 图像翻转

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
```

```python
src = cv2.imread('../datas/f4.jpg')
src = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
dst_x = cv2.flip(src,0) # X轴翻转
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst_x)
plt.title('Flip->X')
```

```python
dst_y = cv2.flip(src,1)# Y轴翻转
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst_y)
plt.title('Flip->Y')
```

```python
dst_xy = cv2.flip(src,-1)# XY轴翻转
```

```python
plt.subplot(121)
plt.imshow(src)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst_xy)
plt.title('Flip->XY')
```
