# 对比度拉伸

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
src = cv2.imread('../datas/f4.jpg')
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
rgb_f = rgb.astype(np.float32)
```

```python
def contrast_stretch(src):
    srcf = src.astype(np.uint8)
    mmin = np.min(srcf)
    mmax = np.max(srcf)
    eps = 1.0e-5
    dst = (srcf - mmin) / (mmax - mmin) * 255
    dst[dst < 0] = 0
    dst[dst > 255] = 255
    dst = dst.astype(np.uint8)
    return dst
```

```python
dst = contrast_stretch(gray)
plt.imshow(dst,'gray')
```
