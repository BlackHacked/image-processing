# 灰度比特平面

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
src = cv2.imread('../datas/f2.jpg')
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
```

```python
def gray_bit_plan(src):
   
    bitplans = []
    
    for n in range(1,9):
        dst = np.zeros_like(src)
        pixmin = 2 ** (n - 1)
        pixmax = 2 ** n
        idx = (src >= pixmin) & (src < pixmax)
        dst[idx] = 255
        dst[~idx] = 0
        bitplans.append(dst)

    return bitplans
                
```

```python
bitplans = gray_bit_plan(gray)
```
