# 颜色映射与伪色彩

```python
import numpy as np
import cv2
import random
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
src = cv2.imread('../datas/pluto.jpg')
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
gray = cv2.imread('../datas/pluto.jpg',0)
```

```python
im_color = cv2.applyColorMap(gray, cv2.COLORMAP_JET)
plt.imshow(im_color)
plt.title('COLORMAP:JET')
```

```python
im_color = cv2.applyColorMap(gray, cv2.COLORMAP_AUTUMN)
plt.imshow(im_color)
plt.title('COLORMAP:AUTUMN')
```

```python
im_color = cv2.applyColorMap(gray, cv2.COLORMAP_BONE)
plt.imshow(im_color)
plt.title('COLORMAP:BONE')
```

```python
im_color = cv2.applyColorMap(gray, cv2.COLORMAP_WINTER)
plt.imshow(im_color)
plt.title('COLORMAP:WINTER')
```

```python
im_color = cv2.applyColorMap(gray, cv2.COLORMAP_RAINBOW	)
plt.imshow(im_color)
plt.title('COLORMAP:RAINBOW')
```

```python
im_color = cv2.applyColorMap(gray, cv2.COLORMAP_OCEAN)
plt.imshow(im_color)
plt.title('COLORMAP:OCEAN')
```

```python
im_color = cv2.applyColorMap(gray, cv2.COLORMAP_SUMMER)
plt.imshow(im_color)
plt.title('COLORMAP:SUMMER')
```

```python
im_color = cv2.applyColorMap(gray, cv2.COLORMAP_HSV)
plt.imshow(im_color)
plt.title('COLORMAP:HSV')
```

```python
im_color = cv2.applyColorMap(gray, cv2.COLORMAP_PINK)
plt.imshow(im_color)
plt.title('COLORMAP:PINK')
```

```python
im_color = cv2.applyColorMap(gray, cv2.COLORMAP_HOT)
plt.imshow(im_color)
plt.title('COLORMAP:HOT')
```
