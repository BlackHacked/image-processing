# 高通滤波

```python
import numpy as np
%matplotlib  inline
import cv2
from scipy import ndimage
import matplotlib.pyplot as plt
```

```python
img = cv2.cvtColor(cv2.imread('../datas/sunset.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.imread('../datas/sunset.jpg',0)
```

```python
kernel_3x3 = np.array([[-1, -1, -1],
                   [-1,  8, -1],
                   [-1, -1, -1]])

kernel_5x5 = np.array([[-1, -1, -1, -1, -1],
                       [-1,  1,  2,  1, -1],
                       [-1,  2,  4,  2, -1],
                       [-1,  1,  2,  1, -1],
                       [-1, -1, -1, -1, -1]])

k3_r = ndimage.convolve(img[:,:,0], kernel_3x3)
k5_r = ndimage.convolve(img[:,:,0], kernel_5x5)
k3_g = ndimage.convolve(img[:,:,1], kernel_3x3)
k5_g = ndimage.convolve(img[:,:,1], kernel_5x5)
k3_b = ndimage.convolve(img[:,:,2], kernel_3x3)
k5_b = ndimage.convolve(img[:,:,2], kernel_5x5)

dst_k3 = cv2.merge([k3_r,k3_g,k3_b])
dst_k5 = cv2.merge([k5_r,k5_g,k5_b])

blurred = cv2.GaussianBlur(img, (17,17), 0)
g_hpf = img - blurred
g_hpf[g_hpf < 0] = 0
```

```python
plt.imshow(np.abs(g_hpf))
```

```python
plt.figure()
plt.imshow(dst_k3)
plt.figure()
plt.imshow(dst_k5)
```
