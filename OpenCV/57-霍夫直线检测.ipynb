# 霍夫直线检测

```python
import numpy as np
%matplotlib  inline
import cv2
from scipy import ndimage
import matplotlib.pyplot as plt
```

```python
img = cv2.cvtColor(cv2.imread('../datas/desktop.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.imread('../datas/desktop.jpg',0)
```

```python
plt.imshow(gray,'gray')
```

```python
edges = cv2.Canny(gray, 0, 255, apertureSize = 3)

### Again we use the same rho and theta accuracies
### However, we specify a minimum vote(points along line) of 100
### and min line length of 5 pixels and max gap between lines of 10 pixels
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 200, 5, 10)
print(lines.shape)

for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(img, (x1, y1), (x2, y2), (0, 255, 0), 3)
```

```python
plt.imshow(img)
```
