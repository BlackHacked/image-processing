# LUT查表反色处理

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
src = cv2.imread('../datas/f4.jpg')
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
rgb_f = rgb.astype(np.float32)
```

```python
lut_table = np.zeros((256,),dtype=np.uint8)
for i in range(256):
    lut_table[i] = 255 - i
    
lookup_table = lut_table.copy()#np.zeros_like(lut_table)
dst = cv2.LUT(rgb,lookup_table)
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('LUT')
```
