# 距离变换

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
src = cv2.imread('../datas/f4.jpg')
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
rgb_f = rgb.astype(np.float32)
```

```python
ret_,thresh = cv2.threshold(gray,160,240,cv2.THRESH_BINARY)
dst = cv2.distanceTransform(thresh,cv2.DIST_L2,cv2.DIST_MASK_PRECISE)
```

```python
plt.imshow(thresh,'gray')
```

```python
plt.imshow(dst,'gray')
```
