# 对数变换

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
src = cv2.imread('../datas/f4.jpg')
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
rgb_f = rgb.astype(np.float32)
```

```python
def log_transform(src,c):
    srcf = src.astype(np.float32) + 1
    dst = c * cv2.log(srcf)
    dst = cv2.normalize(dst,0,255,cv2.NORM_MINMAX )
    return cv2.convertScaleAbs(dst * 255)
```

```python
dst = log_transform(gray,4)
plt.imshow(dst,'gray')
```
