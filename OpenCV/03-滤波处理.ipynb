# 滤波处理

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
```

```python
src = cv2.imread('../datas/f4.jpg')
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
```

## 1.均值滤波

```python
dst = cv2.blur(rgb,(5,5))
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Average Blur')
```

## 2.方框滤波

```python
dst = cv2.boxFilter(rgb,-1,(5,5))
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Box Blur')
```

## 3.高斯滤波

```python
dst = cv2.GaussianBlur(rgb,(5,5),2.4)
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Gaussian Blur')
```

## 4.中值滤波

```python
dst = cv2.medianBlur(rgb,5)
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Median Blur')
```

## 5.双边滤波

```python
dst = cv2.bilateralFilter(rgb,9,75,75)
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('BilateralFilter Blur')
```
