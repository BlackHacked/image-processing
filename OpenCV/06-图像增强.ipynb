# 图像增强

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
src = cv2.imread('../datas/f4.jpg')
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(src,cv2.COLOR_BGR2GRAY)
```

```python
channels = cv2.split(rgb)
```

## 1.直方图均衡化

```python
gr = cv2.equalizeHist(channels[0])
gg = cv2.equalizeHist(channels[1])
gb = cv2.equalizeHist(channels[2])
dst_histeq = cv2.merge([gr,gg,gb])
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst_histeq)
plt.title('Histogram Equation')
```

## 2.对比度

```python
alpha = -1
beta = 4
rgb_f = rgb.astype(np.float64)
dst_contrast = alpha * rgb_f + beta
# dst_contrast[dst_contrast < 0] = 0
dst_contrast = np.abs(dst_contrast)
dst_contrast[dst_contrast > 255] = 255
dst_contrast = dst_contrast.astype(np.uint8)
```

```python
plt.subplot(121)
plt.imshow(rgb)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst_contrast)
plt.title('Contrast')
```
