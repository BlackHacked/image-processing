# 图像滤波-boxFilter


```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
import imageio
```

```python
src = imageio.imread('../../../datas/f4.jpg')
```

```python
help(cv2.boxFilter)
```

```python
dst = cv2.boxFilter(src,-1,(7,7))
```

```python
plt.figure()
plt.imshow(src)
plt.title('Source')
plt.figure()
plt.imshow(dst)
plt.title('Result')
```
