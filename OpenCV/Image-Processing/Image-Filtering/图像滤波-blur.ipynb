# 图像滤波-blur


```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
import imageio
```

```python
src = imageio.imread('../../../datas/face3.jpg')
```

```python
help(cv2.blur)
```

```python
dst = cv2.blur(src,(7,7))
```

```python
plt.figure()
plt.imshow(src)
plt.title('Source')
plt.figure()
plt.imshow(dst)
plt.title('Result')
```
