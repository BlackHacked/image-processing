# 图像滤波-bilateralFilter


```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
import imageio
```

```python
src = imageio.imread('../../../datas/face3.jpg')
```

```python
help(cv2.bilateralFilter)
```

```python
dst = cv2.bilateralFilter(src,7,2.4,2.4)
```

```python
plt.figure()
plt.imshow(src)
plt.title('Source')
plt.figure()
plt.imshow(dst)
plt.title('Result')
```
