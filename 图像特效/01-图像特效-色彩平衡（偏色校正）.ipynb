# 图像特效-色彩平衡（偏色校正）

```python
#coding:utf-8
%matplotlib  inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.imread('../datas/f2.jpg',0)
```

```python
def color_balance(src,r,g,b):
    srci = src.astype(np.int32)
    dst = np.zeros_like(srci)
    
    if r < -255:
        r = -255
    if r > 255:
        r = 255
    if g < -255:
        g = -255
    if g > 255:
        g = 255
    if b < -255:
        b = -255
    if b > 255:
        b = 255
    dst[:,:,0] = src[:,:,0] + r
    dst[:,:,1] = src[:,:,1] + g
    dst[:,:,2] = src[:,:,2] + b
#     dst[dst < 0] = 0
#     dst[dst > 255] = 255
#     dst = np.abs(dst)
#     dst[dst > 255] = 255
    return cv2.convertScaleAbs(dst)
```

```python
dst = color_balance(src,-255,-10,0)
plt.imshow(dst)
```
