# 图像特效-马赛克效果

```python
from scipy import signal
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
```

```python
src = cv2.imread('../datas/face2.jpg')
gray = cv2.imread('../datas/face2.jpg',0)
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
def mosic(src,rsize = 15):
    srcf = np.float32(src)
    if len(src.shape) < 3:
        srcf = cv2.cvtColor(src,cv2.COLOR_GRAY2RGB)
    dst = np.zeros_like(srcf)
    rows,cols = srcf.shape[0],srcf.shape[1]
    offset = (rsize - 1) // 2
    for i in range(offset,rows - offset):
        for j in range(offset,cols - offset):
            for m in range(-offset,offset):
                for n in range(-offset,offset):
                    dst[i + m,j + n] = srcf[i,j]
    return np.uint8(dst)
```

```python
dst = mosic(rgb)
plt.imshow(dst)
```
