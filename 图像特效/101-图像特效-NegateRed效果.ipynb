# 图像特效-NegateRed效果

```python
from scipy import signal
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
src = cv2.imread('../datas/f4.jpg')
gray = cv2.imread('../datas/f4.jpg',0)
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
def negate_red(src):
    dst = src.copy()
    dst[:,:,0] = 255 - src[:,:,0]
    return dst
```

```python
dst = negate_red(rgb)
plt.imshow(dst)
```
