# 图像特效-Acid滤镜

```python
from scipy import signal
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
src = cv2.imread('../datas/f4.jpg')
gray = cv2.imread('../datas/f4.jpg',0)
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
def acid_filter(src,mode='red'):
    dst = np.zeros_like(src).astype(np.float32)
    srcf = np.float32(src)
    if mode == 'red':
        red_avg = np.mean(srcf[:,:,0])
        dst[:,:,0] = red_avg
        dst[:,:,1] = srcf[:,:,1]
        dst[:,:,2] = srcf[:,:,2]
    elif mode == 'green':
        green_avg = np.mean(srcf[:,:,1])
        dst[:,:,0] = srcf[:,:,0]
        dst[:,:,1] = green_avg
        dst[:,:,2] = srcf[:,:,2]
    elif mode == 'blue':
        blue_avg = np.mean(srcf[:,:,2])
        dst[:,:,0] = srcf[:,:,0]
        dst[:,:,1] = srcf[:,:,1]
        dst[:,:,2] = blue_avg
    return np.uint8(dst)
```

```python
dst = acid_filter(rgb)
plt.imshow(dst)
```

```python
dst = acid_filter(rgb,'green')
plt.imshow(dst)
```

```python
dst = acid_filter(rgb,'blue')
plt.imshow(dst)
```
