# 图像特效-雾化

```python
from scipy import signal
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
from numpy.lib.function_base import histogram
```

```python
src = cv2.imread('../datas/sunset2.jpg')
gray = cv2.imread('../datas/sunset2.jpg',0)
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
def fog(src,blocksize):
    srcf = np.float32(src)
    if len(src.shape) < 3:
        srcf = cv2.cvtColor(src,cv2.COLOR_GRAY2RGB).astype(np.float32)
    dst = np.zeros_like(srcf)
    rows,cols = srcf.shape[0],srcf.shape[1]
    for i in range(rows):
        for j in range(cols):
            k = np.random.randint(0,blocksize)
            dx = j + k % blocksize
            dy = i + k % blocksize
            if dx >= cols:
                dx = cols - 1
            if dy >= rows:
                dy = rows - 1
            dst[i,j] = src[dy,dx]
    return np.uint8(dst)
```

```python
dst = fog(rgb,25)
plt.imshow(dst)
```
