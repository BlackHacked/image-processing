# 图像特效-图像错切

```python
#coding:utf-8
%matplotlib  inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
from numpy.lib.function_base import histogram
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.imread('../datas/f2.jpg',0)
```

```python
def slant_x(src,angle):
    srci = src
    if len(src.shape) < 3:
        srci = cv2.cvtColor(src,cv2.COLOR_GRAY2RGB)
    ftan = np.abs(np.tan(angle/180.0*np.pi))
    rows,cols = srci.shape[0],srci.shape[1]
    
    nrows = rows
    ncols = int(cols + rows * ftan)
    dst = np.zeros((nrows,ncols,srci.shape[2]))
    for i in range(nrows):
        for j in range(ncols):
            x0 = i
            y0 = int(j + ftan*(i-rows))
            if x0 >= 0 and x0 < cols and y0 >=0 and y0 < rows:
                dst[i,j] = srci[y0,x0]
    return dst

def slant_y(src,angle):
    srci = src
    if len(src.shape) < 3:
        srci = cv2.cvtColor(src,cv2.COLOR_GRAY2RGB)
    ftan = np.abs(np.tan(angle/180.0*np.pi))
    rows,cols = srci.shape[0],srci.shape[1]
    
    nrows = int(rows + cols * ftan)
    ncols = cols
    dst = np.zeros((nrows,ncols,srci.shape[2]),dtype=src.dtype)
    for i in range(nrows):
        for j in range(ncols):
            x0 = i
            y0 = int(j + ftan*(i-rows))
            if x0 >= 0 and x0 < cols and y0 >=0 and y0 < rows:
                dst[i,j] = srci[y0,x0]
    return dst
```

```python
dst =slant_x(255-src,30)
plt.imshow(dst)
plt.title('Slaut:X')
```

```python
dst =slant_x(255-src,30)
plt.imshow(dst)
plt.title('Slaut:Y')
```
