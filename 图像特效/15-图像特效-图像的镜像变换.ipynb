# 图像特效-图像的镜像变换

```python
#coding:utf-8
%matplotlib  inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
from numpy.lib.function_base import histogram
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.imread('../datas/f2.jpg',0)
```

```python
def mirror_x(src):
    srci = src.copy()
    if len(src.shape) < 3:
        srci = cv2.cvtColor(src,cv2.COLOR_GRAY2RGB)
    dst = np.zeros_like(srci)
    rows,cols = srci.shape[0],srci.shape[1]
    idx = np.arange(0,cols-1)
    idx_r = np.arange(cols-1,0,-1)
    dst[:,idx_r,:] = srci[:,idx,:]
    return dst

def mirror_y(src):
    srci = src.copy()
    if len(src.shape) < 3:
        srci = cv2.cvtColor(src,cv2.COLOR_GRAY2RGB)
    dst = np.zeros_like(srci)
    rows,cols = srci.shape[0],srci.shape[1]
    idy = np.arange(0,rows-1)
    idy_r = np.arange(rows-1,0,-1)
    dst[idy_r,:,:] = srci[idy,:,:]
    return dst

def mirror_xy(src):
    srci = src.copy()
    if len(src.shape) < 3:
        srci = cv2.cvtColor(src,cv2.COLOR_GRAY2RGB)
    dst = np.zeros_like(srci)
    rows,cols = srci.shape[0],srci.shape[1]
    idx = np.arange(0,cols-1).reshape((cols-1,1))
    idx_r = np.arange(cols-1,0,-1).reshape((cols-1,1))
    idy = np.arange(0,rows-1).reshape((1,rows-1))
    idy_r = np.arange(rows-1,0,-1).reshape((1,rows-1))
    dst[idy_r,idx_r] = srci[idy,idx]
    return dst
```

```python
dst = mirror_x(src)
plt.subplot(121)
plt.imshow(src)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Mirror X')
```

```python
dst = mirror_y(src)
plt.subplot(121)
plt.imshow(src)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Mirror Y')
```

```python
dst = mirror_xy(src)
plt.subplot(121)
plt.imshow(src)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Mirror XY')
```
