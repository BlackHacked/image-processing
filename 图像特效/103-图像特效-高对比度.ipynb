# 图像特效-高对比度

```python
from scipy import signal
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
src = cv2.imread('../datas/f4.jpg')
gray = cv2.imread('../datas/f4.jpg',0)
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
def high_contrast(src):
    dst = src.copy()
    idx = src > 128
    dst[idx] = 255
    dst[~idx] = src[~idx]
    return dst
```

```python
dst = high_contrast(rgb)
plt.imshow(dst)
```
