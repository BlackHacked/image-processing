# 图像特效-彩色通道分离与轮换

```python
#coding:utf-8
%matplotlib  inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f3.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.imread('../datas/f2.jpg',0)
```

```python
def rotate_channel(src):
    dst = np.zeros_like(src)
    dst[:,:,0] = src[:,:,1]
    dst[:,:,1] = src[:,:,2]
    dst[:,:,2] = src[:,:,0]
    return dst
```

```python
def split_channel(src):
    return src[:,:,0].copy(),src[:,:,1].copy(),src[:,:,2].copy()
```

```python
dst = rotate_channel(src)
plt.imshow(dst)
plt.title('Rotate RGB Channel')
```

```python
r,g,b = split_channel(src)
plt.subplot(131)
plt.imshow(r,'gray')
plt.title('Channel Red')
plt.subplot(132)
plt.imshow(g,'gray')
plt.title('Channel Green')
plt.subplot(133)
plt.imshow(b,'gray')
plt.title('Channel Blue')
```
