# 图像特效-CarnivalMirror

```python
from scipy import signal
import numpy as np
import cv2
import matplotlib.pyplot as plt
from numpy.lib.function_base import histogram
```

```python
src = cv2.imread('../datas/f2.jpg')
gray = cv2.imread('../datas/f2.jpg',0)
src = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
def carniva_mirror_filter(src,xWavelength,xAmount,yWavelength,yAmount):
    srcf = src.astype(np.float32)
    dst = np.zeros_like(srcf)
    rows,cols = srcf.shape[0],srcf.shape[1]
    for i in range(rows):
        for j in range(cols):
            y = int(i + np.sin(i / yWavelength) * yAmount)
            x = int(j + np.sin(j / xWavelength) * xAmount)
            if y < rows and y >=0 and x < cols and x >=0:
                dst[y,x] = srcf[i,j]
#                 pass
#             else:
#                 dst[i,j] = src[i,j]
    dst = dst.astype(np.uint8)
    return dst
```

```python
dst = carniva_mirror_filter(src,10,20,10,20)
```
