# 图像特效-乘除法

```python
#coding:utf-8
%matplotlib  inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
from numpy.lib.function_base import histogram
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.imread('../datas/f2.jpg',0)

src2 = cv2.cvtColor(cv2.imread('../datas/f1.jpg'),cv2.COLOR_BGR2RGB)
gray2 = cv2.imread('../datas/f1.jpg',0)
```

```python
def multiply(src1,src2):
    dst = np.zeros_like(src1).astype(np.float32)
    dst = src1.astype(np.float32) * src2.astype(np.float32) / 255.0
    return dst.astype(np.uint8)

def division(src1,src2):
    dst = np.zeros_like(src1).astype(np.float32)
    dst = src1.astype(np.float32) / (src2.astype(np.float32)+1.0e-5) * 255
    dst[dst > 255] = 255
    return dst.astype(np.uint8)
```

```python
dst = multiply(src,src2)
plt.imshow(dst)
plt.title('Multiply')
```

```python
dst = division(src,src2)
plt.imshow(dst)
plt.title('Division')
```
