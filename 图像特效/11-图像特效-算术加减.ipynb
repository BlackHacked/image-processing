# 图像特效-算术加减

```python
#coding:utf-8
%matplotlib  inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
from numpy.lib.function_base import histogram
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f1.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.imread('../datas/f2.jpg',0)

src2 = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
gray2 = cv2.imread('../datas/f4.jpg',0)
```

```python
def add(src1,src2):
    dst = np.zeros_like(src1).astype(np.int32)
    dst = src1.astype(np.int32) + src2.astype(np.int32)
    dst[dst > 255] = 255
    dst = dst.astype(np.uint8)
    return dst

def add_weighted(src1,alpha,src2):
    if alpha > 1:
        alpha = 1
    beta = 1.0 - alpha
    dst = np.zeros_like(src1).astype(np.float32)
    dst = alpha * src1 + beta * src2
    dst[dst < 0] = 0
    dst[dst > 255] = 255
    dst = dst.astype(np.uint8)
    return dst

def sub(src1,src2):
    dst = np.zeros_like(src1).astype(np.int32)
    dst = src1.astype(np.int32) - src2.astype(np.int32)
    dst[dst < 0] = 0
    return dst.astype(np.uint8)
```

```python
dst = add(src,src2)
plt.imshow(dst)
plt.title('Add')
```

```python
dst = add_weighted(src,0.75,src2)
plt.imshow(dst)
plt.title('Add Weighted')
```

```python
dst = sub(src,src2)
plt.imshow(dst)
plt.title('Sub')
```
