# 图像特效-Min-Max

```python
#coding:utf-8
%matplotlib  inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
from numpy.lib.function_base import histogram
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f1.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.imread('../datas/f2.jpg',0)

src2 = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
gray2 = cv2.imread('../datas/f4.jpg',0)
```

```python
def min_max(src1,src2,flag='max'):
    if flag == 'max':
        return np.maximum(src1,src2)
    else:
        return np.minimum(src1,src2)
```

```python
dst = min_max(src,src2)
plt.imshow(dst)
plt.title('Max')
```

```python
dst = min_max(src,src2,flag='min')
plt.imshow(dst)
plt.title('Min')
```
