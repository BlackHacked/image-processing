# 图像特效-反转（负片）

```python
#coding:utf-8
%matplotlib  inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f4.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.imread('../datas/f2.jpg',0)
```

```python
def invert(src):
    dst = np.zeros_like(src)
    dst = 255 - src
    return dst
```

```python
dst = invert(src)
plt.subplot(121)
plt.imshow(src)
plt.title('SRC')
plt.subplot(122)
plt.imshow(dst)
plt.title('Invert')
```
