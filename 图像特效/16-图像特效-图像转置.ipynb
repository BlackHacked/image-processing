# 图像特效-图像转置

```python
#coding:utf-8
%matplotlib  inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
from numpy.lib.function_base import histogram
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.imread('../datas/f2.jpg',0)
```

```python
def transpose(src):
    srci = src
    if len(src.shape) < 3:
        srci = cv2.cv2Color(src,cv2.COLOR_GRAY2RGB)
    dst = np.zeros((srci.shape[1],srci.shape[0],srci.shape[2]),dtype=srci.dtype)
    dst[:,:,0] = srci[:,:,0].T
    dst[:,:,1] = srci[:,:,1].T
    dst[:,:,2] = srci[:,:,2].T
    return dst
```

```python
dst = transpose(src)
plt.imshow(dst)
```
