# Sepia颜色效果

```python
from scipy import signal
import numpy as np
import cv2
import matplotlib.pyplot as plt
from numpy.lib.function_base import histogram
```

```python
src = cv2.imread('../datas/f2.jpg')
gray = cv2.imread('../datas/f2.jpg',0)
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
def sepia_filter(img):
    sepia_filter = np.array([[.393, .769, .189],
                             [.349, .686, .168],
                             [.272, .534, .131]])
    scale = 0.01
    sepia_img = img.dot(sepia_filter.T * scale)
    sepia_img /= sepia_img.max()
    sepia_img = sepia_img * 255
    sepia_img[sepia_img < 0] = 0
    sepia_img[sepia_img > 255] = 255
    return sepia_img.astype(np.uint8)
```

```python
dst = sepia_filter(rgb)
plt.imshow(dst)
```
