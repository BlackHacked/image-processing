# 图像特效-曝光过度

```python
from scipy import signal
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
src = cv2.imread('../datas/scenery.jpg')
gray = cv2.imread('../datas/scenery.jpg',0)
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
def solarize(src):
    srcf = np.float32(src)
    if len(src.shape) < 3:
        srcf = cv2.cvtColor(src,cv2.COLOR_GRAY2RGB).astype(np.float32)
    dst = np.zeros_like(srcf)
    idx = srcf > 127
    dst[idx] = 255 - srcf[idx]
    dst[~idx] = srcf[~idx]
    dst = np.clip(dst,0,255)
    return np.uint8(dst)
    
```

```python
dst = solarize(rgb)
plt.imshow(dst)
```
