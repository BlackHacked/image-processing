# 图像特效-暗调滤镜

```python
from scipy import signal
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
src = cv2.imread('../datas/sunset2.jpg')
gray = cv2.imread('../datas/sunset2.jpg',0)
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
def darkness(src):
    srcf = np.float32(src)
    dst = np.zeros_like(srcf)
    dst[:,:,0] = srcf[:,:,0] * srcf[:,:,0] / 255.0
    dst[:,:,1] = srcf[:,:,1] * srcf[:,:,1] / 255.0
    dst[:,:,2] = srcf[:,:,2] * srcf[:,:,2] / 255.0
    dst = np.clip(dst,0,255)
    return np.uint8(dst)
```

```python
dst = darkness(rgb)
plt.imshow(dst)
```
