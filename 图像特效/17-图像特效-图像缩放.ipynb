# 图像特效-图像缩放

```python
#coding:utf-8
%matplotlib  inline
import numpy as np
import cv2
import matplotlib.pyplot as plt
from numpy.lib.function_base import histogram
```

```python
src = cv2.cvtColor(cv2.imread('../datas/f2.jpg'),cv2.COLOR_BGR2RGB)
gray = cv2.imread('../datas/f2.jpg',0)
```

```python
def resize(src,w,h):
    return cv2.resize(src,(w,h))
```

```python
dst = resize(src,src.shape[1] * 2,src.shape[0] * 2)
plt.imshow(dst)
```
