# 图像特效-连环画滤镜

```python
from scipy import signal
import numpy as np
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
src = cv2.imread('../datas/face2.jpg')
gray = cv2.imread('../datas/face2.jpg',0)
rgb = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
```

```python
def cartoon(src):
    srcf = np.float32(src)
    r,g,b = srcf[:,:,0],srcf[:,:,1],srcf[:,:,2]
    R = np.abs(2 * g - b + r)  / 255.0
    G = np.abs(2 * b - g + r) / 255.0
    B = np.abs(2 * b - g + r) / 255.0
    
    gray = (R + G + B).astype(np.float32) / 3.0 * 255
    dst = np.zeros_like(srcf).astype(np.float32)
    dst[:,:,0] = gray + 10
    dst[:,:,1] = gray + 10
    dst[:,:,2] = gray
    dst = np.clip(dst,0,255)
    return np.uint8(dst)
```

```python
dst = cartoon(rgb)
plt.imshow(dst)
```
